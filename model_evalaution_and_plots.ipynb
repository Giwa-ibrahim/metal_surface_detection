```python
import torch
import torch.nn as nn
from torch.utils.data import DataLoader
from torchvision import datasets, transforms
import numpy as np
from sklearn.metrics import classification_report, confusion_matrix, precision_recall_fscore_support
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm

# Set device
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(f"Using device: {device}")

# Define your TinyVGG model architecture
class TinyVGG_v2(nn.Module):
    def __init__(self, input_shape: int, hidden_units: int, output_shape: int) -> None:
        super().__init__()
        self.conv_block_1 = nn.Sequential(
            nn.Conv2d(in_channels=input_shape, out_channels=hidden_units, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.Conv2d(in_channels=hidden_units, out_channels=hidden_units, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2, stride=2),
        )
        self.conv_block_2 = nn.Sequential(
            nn.Conv2d(hidden_units, hidden_units, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.Conv2d(hidden_units, hidden_units, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2, stride=2),
        )
        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Linear(in_features=hidden_units * 64 * 64, out_features=output_shape),
        )

    def forward(self, x: torch.Tensor) -> torch.Tensor:
        x = self.conv_block_1(x)
        x = self.conv_block_2(x)
        x = self.classifier(x)
        return x



# Data loading setup
train_data_path = "/content/drive/MyDrive/Metal Detection Data/Processed_Data/train"
valid_data_path = "/content/drive/MyDrive/Metal Detection Data/Processed_Data/valid"
test_data_path = "/content/drive/MyDrive/Metal Detection Data/Processed_Data/test"

transform = transforms.Compose([
    transforms.Grayscale(num_output_channels=3),
    transforms.Resize((256, 256)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5])
])

train_data = datasets.ImageFolder(train_data_path, transform=transform)
valid_data = datasets.ImageFolder(valid_data_path, transform=transform)
test_data = datasets.ImageFolder(test_data_path, transform=transform)

train_loader = DataLoader(train_data, batch_size=32, shuffle=False)  # No shuffle for evaluation
valid_loader = DataLoader(valid_data, batch_size=32, shuffle=False)
test_loader = DataLoader(test_data, batch_size=32, shuffle=False)

class_names = train_data.classes
class_dict = train_data.class_to_idx
print(f"Classes: {class_names}")
print(f"Class to index mapping: {class_dict}")




# Load your trained model
model_path = "/content/drive/MyDrive/Metal Detection Data/results/adaptive/goa_adaptive_tinyvgg_model.pth"
weights_path = "/content/drive/MyDrive/Metal Detection Data/results/adaptive/my_goa_adaptive_tinyvgg_weights.pth"
complete_path = "/content/drive/MyDrive/Metal Detection Data/results/adaptive/my_goa_adaptive_tinyvgg_complete.pkl"

# Method 1: Try loading complete model first
try:
    model = torch.load(model_path, map_location=device)
    print("✓ Loaded complete model successfully")
except:
    # Method 2: Try loading from pickle file
    try:
        import pickle
        with open(complete_path, 'rb') as f:
            saved_data = pickle.load(f)
            model = saved_data['model'].to(device)
        print("✓ Loaded model from pickle successfully")
    except:
        # Method 3: Load weights with correct architecture
        try:
            # First, let's check what the actual model parameters are
            checkpoint = torch.load(weights_path, map_location=device)

            # Check the shape of the first conv layer to determine hidden_units
            first_conv_weight = checkpoint['conv_block_1.0.weight']
            hidden_units = first_conv_weight.shape[0]  # This will be 128 based on error

            print(f"Detected hidden_units: {hidden_units}")

            # Check classifier structure
            classifier_keys = [k for k in checkpoint.keys() if 'classifier' in k]
            print(f"Classifier keys found: {classifier_keys}")

            # Create model with correct architecture
            class TinyVGG_Corrected(nn.Module):
                def __init__(self, input_shape: int, hidden_units: int, output_shape: int) -> None:
                    super().__init__()
                    self.conv_block_1 = nn.Sequential(
                        nn.Conv2d(in_channels=input_shape, out_channels=hidden_units, kernel_size=3, stride=1, padding=1),
                        nn.ReLU(),
                        nn.Conv2d(in_channels=hidden_units, out_channels=hidden_units, kernel_size=3, stride=1, padding=1),
                        nn.ReLU(),
                        nn.MaxPool2d(kernel_size=2, stride=2),
                    )
                    self.conv_block_2 = nn.Sequential(
                        nn.Conv2d(hidden_units, hidden_units, kernel_size=3, padding=1),
                        nn.ReLU(),
                        nn.Conv2d(hidden_units, hidden_units, kernel_size=3, padding=1),
                        nn.ReLU(),
                        nn.MaxPool2d(kernel_size=2, stride=2),
                    )
                    self.classifier = nn.Sequential(
                        nn.Flatten(),
                        nn.Dropout(0.5),  # This might be the missing layer
                        nn.Linear(in_features=hidden_units * 64 * 64, out_features=output_shape),
                    )

                def forward(self, x: torch.Tensor) -> torch.Tensor:
                    x = self.conv_block_1(x)
                    x = self.conv_block_2(x)
                    x = self.classifier(x)
                    return x

            model = TinyVGG_Corrected(
                input_shape=3,
                hidden_units=hidden_units,
                output_shape=len(class_names)
            ).to(device)

            model.load_state_dict(checkpoint)
            print("✓ Loaded model with corrected architecture")

        except Exception as e:
            print(f"Error loading model with all methods: {e}")
            print("Let's try to load and inspect the model structure...")

            # Last resort: load and check the actual saved model structure
            try:
                saved_model = torch.load(model_path, map_location=device)
                print(f"Model type: {type(saved_model)}")
                if hasattr(saved_model, 'state_dict'):
                    print("Model has state_dict method")
                else:
                    model = saved_model
                    print("✓ Direct model loading successful")
            except Exception as final_e:
                print(f"All loading methods failed: {final_e}")
                exit()

model.eval()  # Set to evaluation mode

def evaluate_model_detailed(model, data_loader, dataset_name, class_names):
    """
    Comprehensive evaluation function that returns predictions, true labels, and metrics
    """
    model.eval()
    all_predictions = []
    all_true_labels = []
    correct = 0
    total = 0

    print(f"\n{'='*50}")
    print(f"DETAILED EVALUATION ON {dataset_name.upper()} SET")
    print(f"{'='*50}")

    with torch.no_grad():
        for batch_idx, (inputs, labels) in enumerate(tqdm(data_loader, desc=f"Evaluating {dataset_name}")):
            inputs, labels = inputs.to(device), labels.to(device)

            outputs = model(inputs)
            _, predicted = torch.max(outputs.data, 1)

            total += labels.size(0)
            correct += (predicted == labels).sum().item()

            # Store predictions and true labels
            all_predictions.extend(predicted.cpu().numpy())
            all_true_labels.extend(labels.cpu().numpy())

    # Calculate accuracy
    accuracy = correct / total

    # Calculate precision, recall, F1-score with zero_division handling
    precision, recall, f1, support = precision_recall_fscore_support(
        all_true_labels, all_predictions, average=None, labels=range(len(class_names)), zero_division=0
    )

    # Calculate macro and weighted averages
    macro_precision, macro_recall, macro_f1, _ = precision_recall_fscore_support(
        all_true_labels, all_predictions, average='macro', zero_division=0
    )

    weighted_precision, weighted_recall, weighted_f1, _ = precision_recall_fscore_support(
        all_true_labels, all_predictions, average='weighted', zero_division=0
    )

    # Print results
    print(f"\nACCURACY: {accuracy:.4f} ({accuracy*100:.2f}%)")
    print(f"\nPER-CLASS METRICS:")
    print("-" * 60)
    print(f"{'Class':<15} {'Precision':<10} {'Recall':<10} {'F1-Score':<10} {'Support':<10}")
    print("-" * 60)

    for i, class_name in enumerate(class_names):
        print(f"{class_name:<15} {precision[i]:<10.4f} {recall[i]:<10.4f} {f1[i]:<10.4f} {support[i]:<10}")

    print("-" * 60)
    print(f"{'Macro Avg':<15} {macro_precision:<10.4f} {macro_recall:<10.4f} {macro_f1:<10.4f} {sum(support):<10}")
    print(f"{'Weighted Avg':<15} {weighted_precision:<10.4f} {weighted_recall:<10.4f} {weighted_f1:<10.4f} {sum(support):<10}")

    return all_true_labels, all_predictions, accuracy, precision, recall, f1, support

def plot_confusion_matrix(y_true, y_pred, class_names, dataset_name, save_path=None):
    """
    Plot and optionally save confusion matrix
    """
    cm = confusion_matrix(y_true, y_pred)

    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=class_names, yticklabels=class_names)
    plt.title(f'Confusion Matrix - {dataset_name} Set', fontsize=16, fontweight='bold')
    plt.xlabel('Predicted Label', fontsize=12)
    plt.ylabel('True Label', fontsize=12)
    plt.tight_layout()

    if save_path:
        plt.savefig(f"{save_path}/confusion_matrix_{dataset_name.lower()}.png",
                   dpi=300, bbox_inches='tight')
        print(f"✓ Confusion matrix saved: confusion_matrix_{dataset_name.lower()}.png")

    plt.show()

    return cm

def print_classification_report(y_true, y_pred, class_names, dataset_name):
    """
    Print detailed classification report using sklearn
    """
    print(f"\n{'='*50}")
    print(f"SKLEARN CLASSIFICATION REPORT - {dataset_name.upper()}")
    print(f"{'='*50}")
    print(classification_report(y_true, y_pred, target_names=class_names, digits=4))

# Evaluate on all datasets
results = {}
save_path = "/content/drive/MyDrive/Metal Detection Data/results/adaptive"

# Evaluate Training Set
print("🔍 EVALUATING TRAINING SET...")
train_true, train_pred, train_acc, train_precision, train_recall, train_f1, train_support = evaluate_model_detailed(
    model, train_loader, "Training", class_names)
results['training'] = {
    'accuracy': train_acc,
    'precision': train_precision,
    'recall': train_recall,
    'f1': train_f1,
    'support': train_support,
    'y_true': train_true,
    'y_pred': train_pred
}

# Plot confusion matrix for training set
print("\n📊 GENERATING CONFUSION MATRIX FOR TRAINING SET...")
train_cm = plot_confusion_matrix(train_true, train_pred, class_names, "Training", save_path)

# Print sklearn classification report for training set
print_classification_report(train_true, train_pred, class_names, "Training")

# Check if model is actually working properly on training data
unique_train_predictions = np.unique(train_pred)
print(f"\n⚠️  TRAINING SET MODEL DIAGNOSTIC:")
print(f"Unique predictions: {unique_train_predictions} (classes: {[class_names[i] for i in unique_train_predictions]})")
print(f"Model is predicting {len(unique_train_predictions)} out of {len(class_names)} classes")

# Evaluate Test Set (most important)
print("🔍 EVALUATING TEST SET...")
test_true, test_pred, test_acc, test_precision, test_recall, test_f1, test_support = evaluate_model_detailed(
    model, test_loader, "Test", class_names)
results['test'] = {
    'accuracy': test_acc,
    'precision': test_precision,
    'recall': test_recall,
    'f1': test_f1,
    'support': test_support,
    'y_true': test_true,
    'y_pred': test_pred
}

# Plot confusion matrix for test set
print("\n📊 GENERATING CONFUSION MATRIX FOR TEST SET...")
test_cm = plot_confusion_matrix(test_true, test_pred, class_names, "Test", save_path)

# Print sklearn classification report for test set
print_classification_report(test_true, test_pred, class_names, "Test")

# Check if model is actually working properly
unique_predictions = np.unique(test_pred)
print(f"\n⚠️  MODEL DIAGNOSTIC:")
print(f"Unique predictions: {unique_predictions} (classes: {[class_names[i] for i in unique_predictions]})")
print(f"Model is predicting {len(unique_predictions)} out of {len(class_names)} classes")

if len(unique_predictions) == 1:
    print("🚨 WARNING: Model is only predicting ONE class! This suggests:")
    print("   - Model loading issues (wrong architecture/weights)")
    print("   - Model not properly trained")
    print("   - Severe overfitting")
    print("   - Data preprocessing mismatch")

    # Let's check a few predictions in detail
    print(f"\nFirst 20 predictions: {test_pred[:20]}")
    print(f"First 20 true labels: {test_true[:20]}")

# Print sklearn classification report for test set
print_classification_report(test_true, test_pred, class_names, "Test")

# Evaluate Validation Set
print("\n🔍 EVALUATING VALIDATION SET...")
valid_true, valid_pred, valid_acc, valid_precision, valid_recall, valid_f1, valid_support = evaluate_model_detailed(
    model, valid_loader, "Validation", class_names)
results['validation'] = {
    'accuracy': valid_acc,
    'precision': valid_precision,
    'recall': valid_recall,
    'f1': valid_f1,
    'support': valid_support,
    'y_true': valid_true,
    'y_pred': valid_pred
}

# Plot confusion matrix for validation set
print("\n📊 GENERATING CONFUSION MATRIX FOR VALIDATION SET...")
valid_cm = plot_confusion_matrix(valid_true, valid_pred, class_names, "Validation", save_path)

# Print sklearn classification report for validation set
print_classification_report(valid_true, valid_pred, class_names, "Validation")

# Summary comparison
print(f"\n{'='*80}")
print("FINAL SUMMARY COMPARISON")
print(f"{'='*80}")
print(f"{'Dataset':<12} {'Accuracy':<10} {'Macro F1':<10} {'Weighted F1':<12}")
print("-" * 50)

# Summary comparison
print(f"\n{'='*80}")
print("FINAL SUMMARY COMPARISON")
print(f"{'='*80}")
print(f"{'Dataset':<12} {'Accuracy':<10} {'Macro F1':<10} {'Weighted F1':<12}")
print("-" * 50)

# Define the order for better comparison
dataset_order = ['training', 'validation', 'test']
for dataset_name in dataset_order:
    if dataset_name in results:
        data = results[dataset_name]
        macro_f1 = np.mean(data['f1'])
        # Calculate weighted F1 properly using support information
        if 'support' in data and len(data['support']) > 0:
            weighted_f1 = np.average(data['f1'], weights=data['support'])
        else:
            weighted_f1 = macro_f1  # fallback to macro if no support info
        print(f"{dataset_name.capitalize():<12} {data['accuracy']:<10.4f} {macro_f1:<10.4f} {weighted_f1:<12.4f}")

print(f"\n{'='*80}")
print("📊 PERFORMANCE ANALYSIS:")
print(f"{'='*80}")

if 'training' in results and 'validation' in results and 'test' in results:
    train_acc = results['training']['accuracy']
    val_acc = results['validation']['accuracy']
    test_acc = results['test']['accuracy']

    print(f"Training vs Validation Gap: {train_acc - val_acc:.4f} ({((train_acc - val_acc)*100):.2f}%)")
    print(f"Training vs Test Gap:       {train_acc - test_acc:.4f} ({((train_acc - test_acc)*100):.2f}%)")
    print(f"Validation vs Test Gap:     {val_acc - test_acc:.4f} ({((val_acc - test_acc)*100):.2f}%)")

    if train_acc - val_acc > 0.05:
        print("⚠️  Large train-validation gap suggests OVERFITTING")
    if val_acc - test_acc > 0.05:
        print("⚠️  Large validation-test gap suggests possible data distribution mismatch")
    if train_acc > 0.95 and test_acc < 0.85:
        print("🚨 High training accuracy but lower test accuracy - model may be overfitting")

print(f"{'='*80}")
print("✅ COMPREHENSIVE EVALUATION COMPLETE!")
print("📁 All confusion matrices saved to:", save_path)
```

#### Basic GOA

```python
import torch
import torch.nn as nn
from torch.utils.data import DataLoader
from torchvision import datasets, transforms
import numpy as np
from sklearn.metrics import classification_report, confusion_matrix, precision_recall_fscore_support
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
import json

# Set device
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(f"Using device: {device}")

# Define your TinyVGG model architecture
class TinyVGG_v2(nn.Module):
    def __init__(self, input_shape: int, hidden_units: int, output_shape: int) -> None:
        super().__init__()
        self.conv_block_1 = nn.Sequential(
            nn.Conv2d(in_channels=input_shape, out_channels=hidden_units, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.Conv2d(in_channels=hidden_units, out_channels=hidden_units, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2, stride=2),
        )
        self.conv_block_2 = nn.Sequential(
            nn.Conv2d(hidden_units, hidden_units, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.Conv2d(hidden_units, hidden_units, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2, stride=2),
        )
        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Linear(in_features=hidden_units * 64 * 64, out_features=output_shape),
        )

    def forward(self, x: torch.Tensor) -> torch.Tensor:
        x = self.conv_block_1(x)
        x = self.conv_block_2(x)
        x = self.classifier(x)
        return x

# Data loading setup
train_data_path = "/content/drive/MyDrive/Metal Detection Data/Processed_Data/train"
valid_data_path = "/content/drive/MyDrive/Metal Detection Data/Processed_Data/valid"
test_data_path = "/content/drive/MyDrive/Metal Detection Data/Processed_Data/test"

transform = transforms.Compose([
    transforms.Grayscale(num_output_channels=3),
    transforms.Resize((256, 256)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5])
])

train_data = datasets.ImageFolder(train_data_path, transform=transform)
valid_data = datasets.ImageFolder(valid_data_path, transform=transform)
test_data = datasets.ImageFolder(test_data_path, transform=transform)

train_loader = DataLoader(train_data, batch_size=32, shuffle=False)  # No shuffle for evaluation
valid_loader = DataLoader(valid_data, batch_size=32, shuffle=False)
test_loader = DataLoader(test_data, batch_size=32, shuffle=False)

class_names = train_data.classes
class_dict = train_data.class_to_idx
print(f"Classes: {class_names}")
print(f"Class to index mapping: {class_dict}")

# Load your basic GOA model
model_path = "/content/drive/MyDrive/Metal Detection Data/results/basic/goa_tinyvgg_model.pth"
info_path = "/content/drive/MyDrive/Metal Detection Data/results/basic/goa_model_info.json"

print(f"\n{'='*60}")
print("LOADING BASIC GOA MODEL")
print(f"{'='*60}")

# First, load the model info to understand the architecture
try:
    with open(info_path, 'r') as f:
        model_info = json.load(f)
    print("✓ Model info loaded successfully")
    print(f"Model info keys: {list(model_info.keys())}")

    if 'best_parameters' in model_info:
        best_params = model_info['best_parameters']
        print(f"Best parameters: {best_params}")

        # Extract hidden_units if available
        if 'hidden_units' in best_params:
            optimal_hidden_units = best_params['hidden_units']
        else:
            optimal_hidden_units = 64  # default fallback
    else:
        optimal_hidden_units = 64  # default fallback

    if 'accuracies' in model_info:
        print(f"Reported accuracies: {model_info['accuracies']}")

except Exception as e:
    print(f"Could not load model info: {e}")
    optimal_hidden_units = 64  # default fallback

print(f"Using hidden_units: {optimal_hidden_units}")

# Load the model with multiple strategies
model_loaded = False

# Strategy 1: Try loading complete model with weights_only=False (for PyTorch 2.6+)
try:
    model = torch.load(model_path, map_location=device, weights_only=False)
    print("✓ Strategy 1: Loaded complete model successfully")
    model_loaded = True
except Exception as e:
    print(f"Strategy 1 failed: {e}")

# Strategy 2: Try loading as state dict with detected architecture
if not model_loaded:
    try:
        checkpoint = torch.load(model_path, map_location=device, weights_only=False)

        if isinstance(checkpoint, dict) and 'state_dict' in checkpoint:
            state_dict = checkpoint['state_dict']
        elif isinstance(checkpoint, dict):
            state_dict = checkpoint
        else:
            state_dict = checkpoint.state_dict()

        # Auto-detect hidden_units from first conv layer
        first_conv_key = 'conv_block_1.0.weight'
        if first_conv_key in state_dict:
            detected_hidden_units = state_dict[first_conv_key].shape[0]
            print(f"Detected hidden_units from weights: {detected_hidden_units}")
        else:
            detected_hidden_units = optimal_hidden_units

        # Create model with detected architecture
        model = TinyVGG_v2(
            input_shape=3,
            hidden_units=detected_hidden_units,
            output_shape=len(class_names)
        ).to(device)

        model.load_state_dict(state_dict)
        print(f"✓ Strategy 2: Loaded model with detected hidden_units={detected_hidden_units}")
        model_loaded = True

    except Exception as e:
        print(f"Strategy 2 failed: {e}")

# Strategy 3: Try with optimal_hidden_units from model info
if not model_loaded:
    try:
        model = TinyVGG_v2(
            input_shape=3,
            hidden_units=optimal_hidden_units,
            output_shape=len(class_names)
        ).to(device)

        checkpoint = torch.load(model_path, map_location=device, weights_only=False)
        if isinstance(checkpoint, dict):
            model.load_state_dict(checkpoint)
        else:
            model = checkpoint

        print(f"✓ Strategy 3: Loaded model with optimal_hidden_units={optimal_hidden_units}")
        model_loaded = True

    except Exception as e:
        print(f"Strategy 3 failed: {e}")

# Strategy 4: Try common hidden_units values
if not model_loaded:
    common_hidden_units = [32, 64, 90, 128, 256]
    for hu in common_hidden_units:
        try:
            model = TinyVGG_v2(
                input_shape=3,
                hidden_units=hu,
                output_shape=len(class_names)
            ).to(device)

            checkpoint = torch.load(model_path, map_location=device, weights_only=False)
            model.load_state_dict(checkpoint)
            print(f"✓ Strategy 4: Loaded model with hidden_units={hu}")
            model_loaded = True
            break

        except Exception as e:
            continue

if not model_loaded:
    print("🚨 ERROR: Could not load model with any strategy!")
    exit()

model.eval()  # Set to evaluation mode

def evaluate_model_detailed(model, data_loader, dataset_name, class_names):
    """
    Comprehensive evaluation function that returns predictions, true labels, and metrics
    """
    model.eval()
    all_predictions = []
    all_true_labels = []
    correct = 0
    total = 0

    print(f"\n{'='*50}")
    print(f"DETAILED EVALUATION ON {dataset_name.upper()} SET")
    print(f"{'='*50}")

    with torch.no_grad():
        for batch_idx, (inputs, labels) in enumerate(tqdm(data_loader, desc=f"Evaluating {dataset_name}")):
            inputs, labels = inputs.to(device), labels.to(device)

            outputs = model(inputs)
            _, predicted = torch.max(outputs.data, 1)

            total += labels.size(0)
            correct += (predicted == labels).sum().item()

            # Store predictions and true labels
            all_predictions.extend(predicted.cpu().numpy())
            all_true_labels.extend(labels.cpu().numpy())

    # Calculate accuracy
    accuracy = correct / total

    # Calculate precision, recall, F1-score with zero_division handling
    precision, recall, f1, support = precision_recall_fscore_support(
        all_true_labels, all_predictions, average=None, labels=range(len(class_names)), zero_division=0
    )

    # Calculate macro and weighted averages
    macro_precision, macro_recall, macro_f1, _ = precision_recall_fscore_support(
        all_true_labels, all_predictions, average='macro', zero_division=0
    )

    weighted_precision, weighted_recall, weighted_f1, _ = precision_recall_fscore_support(
        all_true_labels, all_predictions, average='weighted', zero_division=0
    )

    # Print results
    print(f"\nACCURACY: {accuracy:.4f} ({accuracy*100:.2f}%)")
    print(f"\nPER-CLASS METRICS:")
    print("-" * 60)
    print(f"{'Class':<15} {'Precision':<10} {'Recall':<10} {'F1-Score':<10} {'Support':<10}")
    print("-" * 60)

    for i, class_name in enumerate(class_names):
        print(f"{class_name:<15} {precision[i]:<10.4f} {recall[i]:<10.4f} {f1[i]:<10.4f} {support[i]:<10}")

    print("-" * 60)
    print(f"{'Macro Avg':<15} {macro_precision:<10.4f} {macro_recall:<10.4f} {macro_f1:<10.4f} {sum(support):<10}")
    print(f"{'Weighted Avg':<15} {weighted_precision:<10.4f} {weighted_recall:<10.4f} {weighted_f1:<10.4f} {sum(support):<10}")

    return all_true_labels, all_predictions, accuracy, precision, recall, f1, support

def plot_confusion_matrix(y_true, y_pred, class_names, dataset_name, model_type="Basic_GOA", save_path=None):
    """
    Plot and optionally save confusion matrix
    """
    cm = confusion_matrix(y_true, y_pred)

    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=class_names, yticklabels=class_names)
    plt.title(f'Confusion Matrix - {model_type} - {dataset_name} Set', fontsize=16, fontweight='bold')
    plt.xlabel('Predicted Label', fontsize=12)
    plt.ylabel('True Label', fontsize=12)
    plt.tight_layout()

    if save_path:
        filename = f"confusion_matrix_{model_type.lower()}_{dataset_name.lower()}.png"
        plt.savefig(f"{save_path}/{filename}", dpi=300, bbox_inches='tight')
        print(f"✓ Confusion matrix saved: {filename}")

    plt.show()

    return cm

def print_classification_report(y_true, y_pred, class_names, dataset_name, model_type="Basic GOA"):
    """
    Print detailed classification report using sklearn
    """
    print(f"\n{'='*50}")
    print(f"SKLEARN CLASSIFICATION REPORT - {model_type} - {dataset_name.upper()}")
    print(f"{'='*50}")
    print(classification_report(y_true, y_pred, target_names=class_names, digits=4))

# Evaluate on all datasets
results = {}
save_path = "/content/drive/MyDrive/Metal Detection Data/results/basic"

print(f"\n{'='*80}")
print("BASIC GOA MODEL EVALUATION")
print(f"{'='*80}")

# Evaluate Training Set
print("🔍 EVALUATING TRAINING SET...")
train_true, train_pred, train_acc, train_precision, train_recall, train_f1, train_support = evaluate_model_detailed(
    model, train_loader, "Training", class_names)
results['training'] = {
    'accuracy': train_acc,
    'precision': train_precision,
    'recall': train_recall,
    'f1': train_f1,
    'support': train_support,
    'y_true': train_true,
    'y_pred': train_pred
}

# Plot confusion matrix for training set
print("\n📊 GENERATING CONFUSION MATRIX FOR TRAINING SET...")
train_cm = plot_confusion_matrix(train_true, train_pred, class_names, "Training", "Basic_GOA", save_path)

# Print sklearn classification report for training set
print_classification_report(train_true, train_pred, class_names, "Training", "Basic GOA")

# Evaluate Test Set
print("🔍 EVALUATING TEST SET...")
test_true, test_pred, test_acc, test_precision, test_recall, test_f1, test_support = evaluate_model_detailed(
    model, test_loader, "Test", class_names)
results['test'] = {
    'accuracy': test_acc,
    'precision': test_precision,
    'recall': test_recall,
    'f1': test_f1,
    'support': test_support,
    'y_true': test_true,
    'y_pred': test_pred
}

# Plot confusion matrix for test set
print("\n📊 GENERATING CONFUSION MATRIX FOR TEST SET...")
test_cm = plot_confusion_matrix(test_true, test_pred, class_names, "Test", "Basic_GOA", save_path)

# Print sklearn classification report for test set
print_classification_report(test_true, test_pred, class_names, "Test", "Basic GOA")

# Check model diagnostic
unique_predictions = np.unique(test_pred)
print(f"\n⚠️  MODEL DIAGNOSTIC:")
print(f"Unique predictions: {unique_predictions} (classes: {[class_names[i] for i in unique_predictions]})")
print(f"Model is predicting {len(unique_predictions)} out of {len(class_names)} classes")

# Evaluate Validation Set
print("\n🔍 EVALUATING VALIDATION SET...")
valid_true, valid_pred, valid_acc, valid_precision, valid_recall, valid_f1, valid_support = evaluate_model_detailed(
    model, valid_loader, "Validation", class_names)
results['validation'] = {
    'accuracy': valid_acc,
    'precision': valid_precision,
    'recall': valid_recall,
    'f1': valid_f1,
    'support': valid_support,
    'y_true': valid_true,
    'y_pred': valid_pred
}

# Plot confusion matrix for validation set
print("\n📊 GENERATING CONFUSION MATRIX FOR VALIDATION SET...")
valid_cm = plot_confusion_matrix(valid_true, valid_pred, class_names, "Validation", "Basic_GOA", save_path)

# Print sklearn classification report for validation set
print_classification_report(valid_true, valid_pred, class_names, "Validation", "Basic GOA")

# Summary comparison
print(f"\n{'='*80}")
print("BASIC GOA MODEL - FINAL SUMMARY COMPARISON")
print(f"{'='*80}")
print(f"{'Dataset':<12} {'Accuracy':<10} {'Macro F1':<10} {'Weighted F1':<12}")
print("-" * 50)

# Define the order for better comparison
dataset_order = ['training', 'validation', 'test']
for dataset_name in dataset_order:
    if dataset_name in results:
        data = results[dataset_name]
        macro_f1 = np.mean(data['f1'])
        # Calculate weighted F1 properly using support information
        if 'support' in data and len(data['support']) > 0:
            weighted_f1 = np.average(data['f1'], weights=data['support'])
        else:
            weighted_f1 = macro_f1  # fallback to macro if no support info
        print(f"{dataset_name.capitalize():<12} {data['accuracy']:<10.4f} {macro_f1:<10.4f} {weighted_f1:<12.4f}")

print(f"\n{'='*80}")
print("📊 BASIC GOA PERFORMANCE ANALYSIS:")
print(f"{'='*80}")

if 'training' in results and 'validation' in results and 'test' in results:
    train_acc = results['training']['accuracy']
    val_acc = results['validation']['accuracy']
    test_acc = results['test']['accuracy']

    print(f"Training vs Validation Gap: {train_acc - val_acc:.4f} ({((train_acc - val_acc)*100):.2f}%)")
    print(f"Training vs Test Gap:       {train_acc - test_acc:.4f} ({((train_acc - test_acc)*100):.2f}%)")
    print(f"Validation vs Test Gap:     {val_acc - test_acc:.4f} ({((val_acc - test_acc)*100):.2f}%)")

    if train_acc - val_acc > 0.05:
        print("⚠️  Large train-validation gap suggests OVERFITTING")
    if val_acc - test_acc > 0.05:
        print("⚠️  Large validation-test gap suggests possible data distribution mismatch")
    if train_acc > 0.95 and test_acc < 0.85:
        print("🚨 High training accuracy but lower test accuracy - model may be overfitting")

    print(f"\n🔍 MODEL QUALITY ASSESSMENT:")
    if test_acc > 0.9:
        print("✅ EXCELLENT: Test accuracy > 90%")
    elif test_acc > 0.8:
        print("✅ GOOD: Test accuracy > 80%")
    elif test_acc > 0.7:
        print("⚠️  FAIR: Test accuracy > 70%")
    else:
        print("🚨 POOR: Test accuracy < 70%")

print(f"{'='*80}")
print("✅ BASIC GOA MODEL EVALUATION COMPLETE!")
print("📁 All confusion matrices saved to:", save_path)
print(f"{'='*80}")
```

```python
import torch
import torch.nn as nn
from torch.utils.data import DataLoader
from torchvision import datasets, transforms
import numpy as np
from sklearn.metrics import classification_report, confusion_matrix, precision_recall_fscore_support
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
import json
import pickle

# Set device
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(f"Using device: {device}")

# Define your TinyVGG model architecture
class TinyVGG_v2(nn.Module):
    def __init__(self, input_shape: int, hidden_units: int, output_shape: int) -> None:
        super().__init__()
        self.conv_block_1 = nn.Sequential(
            nn.Conv2d(in_channels=input_shape, out_channels=hidden_units, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.Conv2d(in_channels=hidden_units, out_channels=hidden_units, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2, stride=2),
        )
        self.conv_block_2 = nn.Sequential(
            nn.Conv2d(hidden_units, hidden_units, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.Conv2d(hidden_units, hidden_units, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2, stride=2),
        )
        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Linear(in_features=hidden_units * 64 * 64, out_features=output_shape),
        )

    def forward(self, x: torch.Tensor) -> torch.Tensor:
        x = self.conv_block_1(x)
        x = self.conv_block_2(x)
        x = self.classifier(x)
        return x

# Alternative TinyVGG architecture with dropout (in case it's needed)
class TinyVGG_with_Dropout(nn.Module):
    def __init__(self, input_shape: int, hidden_units: int, output_shape: int, dropout_rate: float = 0.5) -> None:
        super().__init__()
        self.conv_block_1 = nn.Sequential(
            nn.Conv2d(in_channels=input_shape, out_channels=hidden_units, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.Conv2d(in_channels=hidden_units, out_channels=hidden_units, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2, stride=2),
        )
        self.conv_block_2 = nn.Sequential(
            nn.Conv2d(hidden_units, hidden_units, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.Conv2d(hidden_units, hidden_units, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2, stride=2),
        )
        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Dropout(dropout_rate),
            nn.Linear(in_features=hidden_units * 64 * 64, out_features=output_shape),
        )

    def forward(self, x: torch.Tensor) -> torch.Tensor:
        x = self.conv_block_1(x)
        x = self.conv_block_2(x)
        x = self.classifier(x)
        return x

# Data loading setup
train_data_path = "/content/drive/MyDrive/Metal Detection Data/Processed_Data/train"
valid_data_path = "/content/drive/MyDrive/Metal Detection Data/Processed_Data/valid"
test_data_path = "/content/drive/MyDrive/Metal Detection Data/Processed_Data/test"

transform = transforms.Compose([
    transforms.Grayscale(num_output_channels=3),
    transforms.Resize((256, 256)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5])
])

train_data = datasets.ImageFolder(train_data_path, transform=transform)
valid_data = datasets.ImageFolder(valid_data_path, transform=transform)
test_data = datasets.ImageFolder(test_data_path, transform=transform)

train_loader = DataLoader(train_data, batch_size=32, shuffle=False)  # No shuffle for evaluation
valid_loader = DataLoader(valid_data, batch_size=32, shuffle=False)
test_loader = DataLoader(test_data, batch_size=32, shuffle=False)

class_names = train_data.classes
class_dict = train_data.class_to_idx
print(f"Classes: {class_names}")
print(f"Class to index mapping: {class_dict}")

# Define model file paths
base_path = "/content/drive/MyDrive/Metal Detection Data/results/levy_flight_2"
model_complete_path = f"{base_path}/goa_levyflight_final_tinyvgg_model_complete.pth"
model_weights_path = f"{base_path}/goa_levyflight_final_tinyvgg_model_weights.pth"
model_pickle_path = f"{base_path}/goa_levyflight_final_tinyvgg_model_complete.pkl"
model_info_path = f"{base_path}/goa_levyflight_final_tinyvgg_model_info.json"

print(f"\n{'='*60}")
print("LOADING LEVY FLIGHT GOA MODEL")
print(f"{'='*60}")

# First, load the model info to understand the architecture
optimal_hidden_units = 64  # default
dropout_rate = 0.5  # default
reported_accuracies = None

try:
    with open(model_info_path, 'r') as f:
        model_info = json.load(f)
    print("✓ Model info loaded successfully")
    print(f"Model info keys: {list(model_info.keys())}")

    if 'best_parameters' in model_info:
        best_params = model_info['best_parameters']
        print(f"Best parameters: {best_params}")

        # Extract parameters
        if 'hidden_units' in best_params:
            optimal_hidden_units = best_params['hidden_units']
        if 'dropout_rate' in best_params:
            dropout_rate = best_params['dropout_rate']

    if 'accuracies' in model_info:
        reported_accuracies = model_info['accuracies']
        print(f"Reported accuracies: {reported_accuracies}")

except Exception as e:
    print(f"Could not load model info: {e}")

print(f"Using hidden_units: {optimal_hidden_units}")
print(f"Using dropout_rate: {dropout_rate}")

# Load the model with multiple strategies
model_loaded = False
loading_strategy = ""

# Strategy 1: Try loading from pickle file (most comprehensive)
try:
    with open(model_pickle_path, 'rb') as f:
        saved_data = pickle.load(f)

    if 'model' in saved_data:
        model = saved_data['model'].to(device)
        print("✓ Strategy 1: Loaded complete model from pickle successfully")
        if 'parameters' in saved_data:
            print(f"  Parameters from pickle: {saved_data['parameters']}")
        if 'accuracies' in saved_data:
            print(f"  Accuracies from pickle: {saved_data['accuracies']}")
        loading_strategy = "pickle"
        model_loaded = True
    else:
        print("Strategy 1: Pickle file doesn't contain 'model' key")

except Exception as e:
    print(f"Strategy 1 failed: {e}")

# Strategy 2: Try loading complete model file
if not model_loaded:
    try:
        model = torch.load(model_complete_path, map_location=device, weights_only=False)
        print("✓ Strategy 2: Loaded complete model from .pth file successfully")
        loading_strategy = "complete_pth"
        model_loaded = True
    except Exception as e:
        print(f"Strategy 2 failed: {e}")

# Strategy 3: Try loading weights with detected architecture
if not model_loaded:
    try:
        checkpoint = torch.load(model_weights_path, map_location=device, weights_only=False)

        # Auto-detect hidden_units from first conv layer
        if 'conv_block_1.0.weight' in checkpoint:
            detected_hidden_units = checkpoint['conv_block_1.0.weight'].shape[0]
            print(f"Detected hidden_units from weights: {detected_hidden_units}")
        else:
            detected_hidden_units = optimal_hidden_units

        # Try with dropout first (as this is Levy Flight which might use dropout)
        try:
            model = TinyVGG_with_Dropout(
                input_shape=3,
                hidden_units=detected_hidden_units,
                output_shape=len(class_names),
                dropout_rate=dropout_rate
            ).to(device)

            model.load_state_dict(checkpoint)
            print(f"✓ Strategy 3a: Loaded model with dropout (hidden_units={detected_hidden_units})")
            loading_strategy = "weights_with_dropout"
            model_loaded = True

        except Exception:
            # Try without dropout
            model = TinyVGG_v2(
                input_shape=3,
                hidden_units=detected_hidden_units,
                output_shape=len(class_names)
            ).to(device)

            model.load_state_dict(checkpoint)
            print(f"✓ Strategy 3b: Loaded model without dropout (hidden_units={detected_hidden_units})")
            loading_strategy = "weights_no_dropout"
            model_loaded = True

    except Exception as e:
        print(f"Strategy 3 failed: {e}")

# Strategy 4: Try with various architectures using optimal parameters
if not model_loaded:
    architectures_to_try = [
        ("TinyVGG_with_Dropout", TinyVGG_with_Dropout, {"dropout_rate": dropout_rate}),
        ("TinyVGG_v2", TinyVGG_v2, {}),
    ]

    hidden_units_to_try = [optimal_hidden_units, 64, 90, 128, 32, 256]

    for arch_name, arch_class, arch_kwargs in architectures_to_try:
        for hu in hidden_units_to_try:
            try:
                model = arch_class(
                    input_shape=3,
                    hidden_units=hu,
                    output_shape=len(class_names),
                    **arch_kwargs
                ).to(device)

                checkpoint = torch.load(model_weights_path, map_location=device, weights_only=False)
                model.load_state_dict(checkpoint)
                print(f"✓ Strategy 4: Loaded {arch_name} with hidden_units={hu}")
                loading_strategy = f"strategy4_{arch_name}_{hu}"
                model_loaded = True
                break

            except Exception:
                continue

        if model_loaded:
            break

if not model_loaded:
    print("🚨 ERROR: Could not load model with any strategy!")
    exit()

model.eval()  # Set to evaluation mode
print(f"Model loaded using: {loading_strategy}")

def evaluate_model_detailed(model, data_loader, dataset_name, class_names):
    """
    Comprehensive evaluation function that returns predictions, true labels, and metrics
    """
    model.eval()
    all_predictions = []
    all_true_labels = []
    correct = 0
    total = 0

    print(f"\n{'='*50}")
    print(f"DETAILED EVALUATION ON {dataset_name.upper()} SET")
    print(f"{'='*50}")

    with torch.no_grad():
        for batch_idx, (inputs, labels) in enumerate(tqdm(data_loader, desc=f"Evaluating {dataset_name}")):
            inputs, labels = inputs.to(device), labels.to(device)

            outputs = model(inputs)
            _, predicted = torch.max(outputs.data, 1)

            total += labels.size(0)
            correct += (predicted == labels).sum().item()

            # Store predictions and true labels
            all_predictions.extend(predicted.cpu().numpy())
            all_true_labels.extend(labels.cpu().numpy())

    # Calculate accuracy
    accuracy = correct / total

    # Calculate precision, recall, F1-score with zero_division handling
    precision, recall, f1, support = precision_recall_fscore_support(
        all_true_labels, all_predictions, average=None, labels=range(len(class_names)), zero_division=0
    )

    # Calculate macro and weighted averages
    macro_precision, macro_recall, macro_f1, _ = precision_recall_fscore_support(
        all_true_labels, all_predictions, average='macro', zero_division=0
    )

    weighted_precision, weighted_recall, weighted_f1, _ = precision_recall_fscore_support(
        all_true_labels, all_predictions, average='weighted', zero_division=0
    )

    # Print results
    print(f"\nACCURACY: {accuracy:.4f} ({accuracy*100:.2f}%)")
    print(f"\nPER-CLASS METRICS:")
    print("-" * 60)
    print(f"{'Class':<15} {'Precision':<10} {'Recall':<10} {'F1-Score':<10} {'Support':<10}")
    print("-" * 60)

    for i, class_name in enumerate(class_names):
        print(f"{class_name:<15} {precision[i]:<10.4f} {recall[i]:<10.4f} {f1[i]:<10.4f} {support[i]:<10}")

    print("-" * 60)
    print(f"{'Macro Avg':<15} {macro_precision:<10.4f} {macro_recall:<10.4f} {macro_f1:<10.4f} {sum(support):<10}")
    print(f"{'Weighted Avg':<15} {weighted_precision:<10.4f} {weighted_recall:<10.4f} {weighted_f1:<10.4f} {sum(support):<10}")

    return all_true_labels, all_predictions, accuracy, precision, recall, f1, support

def plot_confusion_matrix(y_true, y_pred, class_names, dataset_name, model_type="Basic_GOA", save_path=None):
    """
    Plot and optionally save confusion matrix
    """
    cm = confusion_matrix(y_true, y_pred)

    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=class_names, yticklabels=class_names)
    plt.title(f'Confusion Matrix - {model_type} - {dataset_name} Set', fontsize=16, fontweight='bold')
    plt.xlabel('Predicted Label', fontsize=12)
    plt.ylabel('True Label', fontsize=12)
    plt.tight_layout()

    if save_path:
        filename = f"confusion_matrix_{model_type.lower()}_{dataset_name.lower()}.png"
        plt.savefig(f"{save_path}/{filename}", dpi=300, bbox_inches='tight')
        print(f"✓ Confusion matrix saved: {filename}")

    plt.show()

    return cm

def print_classification_report(y_true, y_pred, class_names, dataset_name, model_type="Levy Flight GOA"):
    """
    Print detailed classification report using sklearn
    """
    print(f"\n{'='*50}")
    print(f"SKLEARN CLASSIFICATION REPORT - {model_type} - {dataset_name.upper()}")
    print(f"{'='*50}")
    print(classification_report(y_true, y_pred, target_names=class_names, digits=4))

# Evaluate on all datasets
results = {}
save_path = "/content/drive/MyDrive/Metal Detection Data/results/levy_flight_2"

print(f"\n{'='*80}")
print("LEVY FLIGHT GOA MODEL EVALUATION")
print(f"{'='*80}")

# Show reported accuracies if available
if reported_accuracies:
    print("📊 EXPECTED PERFORMANCE (from model info):")
    for key, value in reported_accuracies.items():
        print(f"  {key}: {value:.4f} ({value*100:.2f}%)")
    print()

# Evaluate Training Set
print("🔍 EVALUATING TRAINING SET...")
train_true, train_pred, train_acc, train_precision, train_recall, train_f1, train_support = evaluate_model_detailed(
    model, train_loader, "Training", class_names)
results['training'] = {
    'accuracy': train_acc,
    'precision': train_precision,
    'recall': train_recall,
    'f1': train_f1,
    'support': train_support,
    'y_true': train_true,
    'y_pred': train_pred
}

# Plot confusion matrix for training set
print("\n📊 GENERATING CONFUSION MATRIX FOR TRAINING SET...")
train_cm = plot_confusion_matrix(train_true, train_pred, class_names, "Training", "Levy_Flight_GOA", save_path)

# Print sklearn classification report for training set
print_classification_report(train_true, train_pred, class_names, "Training", "Levy Flight GOA")

# Evaluate Test Set
print("🔍 EVALUATING TEST SET...")
test_true, test_pred, test_acc, test_precision, test_recall, test_f1, test_support = evaluate_model_detailed(
    model, test_loader, "Test", class_names)
results['test'] = {
    'accuracy': test_acc,
    'precision': test_precision,
    'recall': test_recall,
    'f1': test_f1,
    'support': test_support,
    'y_true': test_true,
    'y_pred': test_pred
}

# Plot confusion matrix for test set
print("\n📊 GENERATING CONFUSION MATRIX FOR TEST SET...")
test_cm = plot_confusion_matrix(test_true, test_pred, class_names, "Test", "Levy_Flight_GOA", save_path)

# Print sklearn classification report for test set
print_classification_report(test_true, test_pred, class_names, "Test", "Levy Flight GOA")

# Check model diagnostic
unique_predictions = np.unique(test_pred)
print(f"\n⚠️  MODEL DIAGNOSTIC:")
print(f"Unique predictions: {unique_predictions} (classes: {[class_names[i] for i in unique_predictions]})")
print(f"Model is predicting {len(unique_predictions)} out of {len(class_names)} classes")

# Evaluate Validation Set
print("\n🔍 EVALUATING VALIDATION SET...")
valid_true, valid_pred, valid_acc, valid_precision, valid_recall, valid_f1, valid_support = evaluate_model_detailed(
    model, valid_loader, "Validation", class_names)
results['validation'] = {
    'accuracy': valid_acc,
    'precision': valid_precision,
    'recall': valid_recall,
    'f1': valid_f1,
    'support': valid_support,
    'y_true': valid_true,
    'y_pred': valid_pred
}

# Plot confusion matrix for validation set
print("\n📊 GENERATING CONFUSION MATRIX FOR VALIDATION SET...")
valid_cm = plot_confusion_matrix(valid_true, valid_pred, class_names, "Validation", "Levy_Flight_GOA", save_path)

# Print sklearn classification report for validation set
print_classification_report(valid_true, valid_pred, class_names, "Validation", "Levy Flight GOA")

# Summary comparison
print(f"\n{'='*80}")
print("LEVY FLIGHT GOA MODEL - FINAL SUMMARY COMPARISON")
print(f"{'='*80}")

# Compare with reported accuracies
if reported_accuracies:
    print("📋 EXPECTED vs ACTUAL PERFORMANCE:")
    print("-" * 50)
    if 'training_accuracy' in reported_accuracies:
        expected_train = reported_accuracies['training_accuracy']
        actual_train = results['training']['accuracy']
        print(f"Training:   Expected {expected_train:.4f} | Actual {actual_train:.4f} | Diff: {actual_train-expected_train:+.4f}")

    if 'validation_accuracy' in reported_accuracies:
        expected_val = reported_accuracies['validation_accuracy']
        actual_val = results['validation']['accuracy']
        print(f"Validation: Expected {expected_val:.4f} | Actual {actual_val:.4f} | Diff: {actual_val-expected_val:+.4f}")

    if 'test_accuracy' in reported_accuracies:
        expected_test = reported_accuracies['test_accuracy']
        actual_test = results['test']['accuracy']
        print(f"Test:       Expected {expected_test:.4f} | Actual {actual_test:.4f} | Diff: {actual_test-expected_test:+.4f}")

    print()

print(f"{'Dataset':<12} {'Accuracy':<10} {'Macro F1':<10} {'Weighted F1':<12}")
print("-" * 50)

# Define the order for better comparison
dataset_order = ['training', 'validation', 'test']
for dataset_name in dataset_order:
    if dataset_name in results:
        data = results[dataset_name]
        macro_f1 = np.mean(data['f1'])
        # Calculate weighted F1 properly using support information
        if 'support' in data and len(data['support']) > 0:
            weighted_f1 = np.average(data['f1'], weights=data['support'])
        else:
            weighted_f1 = macro_f1  # fallback to macro if no support info
        print(f"{dataset_name.capitalize():<12} {data['accuracy']:<10.4f} {macro_f1:<10.4f} {weighted_f1:<12.4f}")

print(f"\n{'='*80}")
print("📊 LEVY FLIGHT GOA PERFORMANCE ANALYSIS:")
print(f"{'='*80}")

if 'training' in results and 'validation' in results and 'test' in results:
    train_acc = results['training']['accuracy']
    val_acc = results['validation']['accuracy']
    test_acc = results['test']['accuracy']

    print(f"Training vs Validation Gap: {train_acc - val_acc:.4f} ({((train_acc - val_acc)*100):.2f}%)")
    print(f"Training vs Test Gap:       {train_acc - test_acc:.4f} ({((train_acc - test_acc)*100):.2f}%)")
    print(f"Validation vs Test Gap:     {val_acc - test_acc:.4f} ({((val_acc - test_acc)*100):.2f}%)")

    if train_acc - val_acc > 0.05:
        print("⚠️  Large train-validation gap suggests OVERFITTING")
    if val_acc - test_acc > 0.05:
        print("⚠️  Large validation-test gap suggests possible data distribution mismatch")
    if train_acc > 0.95 and test_acc < 0.85:
        print("🚨 High training accuracy but lower test accuracy - model may be overfitting")

    print(f"\n🔍 LEVY FLIGHT MODEL QUALITY ASSESSMENT:")
    if test_acc > 0.9:
        print("✅ EXCELLENT: Test accuracy > 90%")
    elif test_acc > 0.8:
        print("✅ GOOD: Test accuracy > 80%")
    elif test_acc > 0.7:
        print("⚠️  FAIR: Test accuracy > 70%")
    else:
        print("🚨 POOR: Test accuracy < 70%")

print(f"\n🚀 LEVY FLIGHT SPECIFIC INSIGHTS:")
print("Levy Flight is known for:")
print("• Enhanced exploration capabilities")
print("• Better global optimization")
print("• Reduced chance of local minima")
print("• Potentially better generalization")

print(f"{'='*80}")
print("✅ LEVY FLIGHT GOA MODEL EVALUATION COMPLETE!")
print("📁 All confusion matrices saved to:", save_path)
print(f"Loading strategy used: {loading_strategy}")
print(f"{'='*80}")
```

```python
import torch
import torch.nn as nn
from torch.utils.data import DataLoader
from torchvision import datasets, transforms
import numpy as np
from sklearn.metrics import classification_report, confusion_matrix, precision_recall_fscore_support
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
import json

# Set device
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(f"Using device: {device}")

# Define your TinyVGG model architecture
class TinyVGG_v2(nn.Module):
    def __init__(self, input_shape: int, hidden_units: int, output_shape: int) -> None:
        super().__init__()
        self.conv_block_1 = nn.Sequential(
            nn.Conv2d(in_channels=input_shape, out_channels=hidden_units, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.Conv2d(in_channels=hidden_units, out_channels=hidden_units, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2, stride=2),
        )
        self.conv_block_2 = nn.Sequential(
            nn.Conv2d(hidden_units, hidden_units, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.Conv2d(hidden_units, hidden_units, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2, stride=2),
        )
        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Linear(in_features=hidden_units * 64 * 64, out_features=output_shape),
        )

    def forward(self, x: torch.Tensor) -> torch.Tensor:
        x = self.conv_block_1(x)
        x = self.conv_block_2(x)
        x = self.classifier(x)
        return x

# Alternative TinyVGG architecture with dropout (in case it's needed)
class TinyVGG_with_Dropout(nn.Module):
    def __init__(self, input_shape: int, hidden_units: int, output_shape: int, dropout_rate: float = 0.5) -> None:
        super().__init__()
        self.conv_block_1 = nn.Sequential(
            nn.Conv2d(in_channels=input_shape, out_channels=hidden_units, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.Conv2d(in_channels=hidden_units, out_channels=hidden_units, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2, stride=2),
        )
        self.conv_block_2 = nn.Sequential(
            nn.Conv2d(hidden_units, hidden_units, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.Conv2d(hidden_units, hidden_units, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2, stride=2),
        )
        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Dropout(dropout_rate),
            nn.Linear(in_features=hidden_units * 64 * 64, out_features=output_shape),
        )

    def forward(self, x: torch.Tensor) -> torch.Tensor:
        x = self.conv_block_1(x)
        x = self.conv_block_2(x)
        x = self.classifier(x)
        return x

# Data loading setup
train_data_path = "/content/drive/MyDrive/Metal Detection Data/Processed_Data/train"
valid_data_path = "/content/drive/MyDrive/Metal Detection Data/Processed_Data/valid"
test_data_path = "/content/drive/MyDrive/Metal Detection Data/Processed_Data/test"

transform = transforms.Compose([
    transforms.Grayscale(num_output_channels=3),
    transforms.Resize((256, 256)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5])
])

train_data = datasets.ImageFolder(train_data_path, transform=transform)
valid_data = datasets.ImageFolder(valid_data_path, transform=transform)
test_data = datasets.ImageFolder(test_data_path, transform=transform)

train_loader = DataLoader(train_data, batch_size=32, shuffle=False)  # No shuffle for evaluation
valid_loader = DataLoader(valid_data, batch_size=32, shuffle=False)
test_loader = DataLoader(test_data, batch_size=32, shuffle=False)

class_names = train_data.classes
class_dict = train_data.class_to_idx
print(f"Classes: {class_names}")
print(f"Class to index mapping: {class_dict}")

# Define model file paths
base_path = "/content/drive/MyDrive/Metal Detection Data/results/pred_aware"
model_path = f"{base_path}/goa_pred_aware_tinyvgg_model.pth"
model_info_path = f"{base_path}/goa_model_info.json"

print(f"\n{'='*60}")
print("LOADING PREDATOR AWARE MOMENTUM GOA MODEL")
print(f"{'='*60}")

# First, load the model info to understand the architecture
optimal_hidden_units = 64  # default
dropout_rate = 0.5  # default
reported_accuracies = None

try:
    with open(model_info_path, 'r') as f:
        model_info = json.load(f)
    print("✓ Model info loaded successfully")
    print(f"Model info keys: {list(model_info.keys())}")

    if 'best_parameters' in model_info:
        best_params = model_info['best_parameters']
        print(f"Best parameters: {best_params}")

        # Extract parameters
        if 'hidden_units' in best_params:
            optimal_hidden_units = best_params['hidden_units']
        if 'dropout_rate' in best_params:
            dropout_rate = best_params['dropout_rate']

    if 'accuracies' in model_info:
        reported_accuracies = model_info['accuracies']
        print(f"Reported accuracies: {reported_accuracies}")

    # Check for model type
    if 'model_type' in model_info:
        print(f"Model type: {model_info['model_type']}")

except Exception as e:
    print(f"Could not load model info: {e}")

print(f"Using hidden_units: {optimal_hidden_units}")
print(f"Using dropout_rate: {dropout_rate}")

# Load the model with multiple strategies
model_loaded = False
loading_strategy = ""

# Strategy 1: Try loading complete model with weights_only=False (for PyTorch 2.6+)
try:
    model = torch.load(model_path, map_location=device, weights_only=False)
    print("✓ Strategy 1: Loaded complete model successfully")
    loading_strategy = "complete_model"
    model_loaded = True
except Exception as e:
    print(f"Strategy 1 failed: {e}")

# Strategy 2: Try loading as state dict and reconstruct model
if not model_loaded:
    try:
        checkpoint = torch.load(model_path, map_location=device, weights_only=False)

        if isinstance(checkpoint, dict) and 'state_dict' in checkpoint:
            state_dict = checkpoint['state_dict']
        elif isinstance(checkpoint, dict):
            state_dict = checkpoint
        else:
            state_dict = checkpoint.state_dict() if hasattr(checkpoint, 'state_dict') else checkpoint

        # Auto-detect hidden_units from first conv layer
        if 'conv_block_1.0.weight' in state_dict:
            detected_hidden_units = state_dict['conv_block_1.0.weight'].shape[0]
            print(f"Detected hidden_units from weights: {detected_hidden_units}")
        else:
            detected_hidden_units = optimal_hidden_units

        # Try with dropout first (Predator Aware might use regularization)
        try:
            model = TinyVGG_with_Dropout(
                input_shape=3,
                hidden_units=detected_hidden_units,
                output_shape=len(class_names),
                dropout_rate=dropout_rate
            ).to(device)

            model.load_state_dict(state_dict)
            print(f"✓ Strategy 2a: Loaded model with dropout (hidden_units={detected_hidden_units})")
            loading_strategy = "weights_with_dropout"
            model_loaded = True

        except Exception:
            # Try without dropout
            model = TinyVGG_v2(
                input_shape=3,
                hidden_units=detected_hidden_units,
                output_shape=len(class_names)
            ).to(device)

            model.load_state_dict(state_dict)
            print(f"✓ Strategy 2b: Loaded model without dropout (hidden_units={detected_hidden_units})")
            loading_strategy = "weights_no_dropout"
            model_loaded = True

    except Exception as e:
        print(f"Strategy 2 failed: {e}")

# Strategy 3: Try with optimal parameters from model info
if not model_loaded:
    try:
        model = TinyVGG_v2(
            input_shape=3,
            hidden_units=optimal_hidden_units,
            output_shape=len(class_names)
        ).to(device)

        checkpoint = torch.load(model_path, map_location=device, weights_only=False)
        if isinstance(checkpoint, dict):
            model.load_state_dict(checkpoint)
        else:
            model = checkpoint

        print(f"✓ Strategy 3: Loaded model with optimal_hidden_units={optimal_hidden_units}")
        loading_strategy = "optimal_params"
        model_loaded = True

    except Exception as e:
        print(f"Strategy 3 failed: {e}")

# Strategy 4: Try multiple architectures and hidden_units combinations
if not model_loaded:
    architectures_to_try = [
        ("TinyVGG_with_Dropout", TinyVGG_with_Dropout, {"dropout_rate": dropout_rate}),
        ("TinyVGG_v2", TinyVGG_v2, {}),
    ]

    hidden_units_to_try = [optimal_hidden_units, 64, 90, 128, 32, 256]

    for arch_name, arch_class, arch_kwargs in architectures_to_try:
        for hu in hidden_units_to_try:
            try:
                model = arch_class(
                    input_shape=3,
                    hidden_units=hu,
                    output_shape=len(class_names),
                    **arch_kwargs
                ).to(device)

                checkpoint = torch.load(model_path, map_location=device, weights_only=False)
                model.load_state_dict(checkpoint)
                print(f"✓ Strategy 4: Loaded {arch_name} with hidden_units={hu}")
                loading_strategy = f"strategy4_{arch_name}_{hu}"
                model_loaded = True
                break

            except Exception:
                continue

        if model_loaded:
            break

if not model_loaded:
    print("🚨 ERROR: Could not load model with any strategy!")
    exit()

model.eval()  # Set to evaluation mode
print(f"Model loaded using: {loading_strategy}")

def evaluate_model_detailed(model, data_loader, dataset_name, class_names):
    """
    Comprehensive evaluation function that returns predictions, true labels, and metrics
    """
    model.eval()
    all_predictions = []
    all_true_labels = []
    correct = 0
    total = 0

    print(f"\n{'='*50}")
    print(f"DETAILED EVALUATION ON {dataset_name.upper()} SET")
    print(f"{'='*50}")

    with torch.no_grad():
        for batch_idx, (inputs, labels) in enumerate(tqdm(data_loader, desc=f"Evaluating {dataset_name}")):
            inputs, labels = inputs.to(device), labels.to(device)

            outputs = model(inputs)
            _, predicted = torch.max(outputs.data, 1)

            total += labels.size(0)
            correct += (predicted == labels).sum().item()

            # Store predictions and true labels
            all_predictions.extend(predicted.cpu().numpy())
            all_true_labels.extend(labels.cpu().numpy())

    # Calculate accuracy
    accuracy = correct / total

    # Calculate precision, recall, F1-score with zero_division handling
    precision, recall, f1, support = precision_recall_fscore_support(
        all_true_labels, all_predictions, average=None, labels=range(len(class_names)), zero_division=0
    )

    # Calculate macro and weighted averages
    macro_precision, macro_recall, macro_f1, _ = precision_recall_fscore_support(
        all_true_labels, all_predictions, average='macro', zero_division=0
    )

    weighted_precision, weighted_recall, weighted_f1, _ = precision_recall_fscore_support(
        all_true_labels, all_predictions, average='weighted', zero_division=0
    )

    # Print results
    print(f"\nACCURACY: {accuracy:.4f} ({accuracy*100:.2f}%)")
    print(f"\nPER-CLASS METRICS:")
    print("-" * 60)
    print(f"{'Class':<15} {'Precision':<10} {'Recall':<10} {'F1-Score':<10} {'Support':<10}")
    print("-" * 60)

    for i, class_name in enumerate(class_names):
        print(f"{class_name:<15} {precision[i]:<10.4f} {recall[i]:<10.4f} {f1[i]:<10.4f} {support[i]:<10}")

    print("-" * 60)
    print(f"{'Macro Avg':<15} {macro_precision:<10.4f} {macro_recall:<10.4f} {macro_f1:<10.4f} {sum(support):<10}")
    print(f"{'Weighted Avg':<15} {weighted_precision:<10.4f} {weighted_recall:<10.4f} {weighted_f1:<10.4f} {sum(support):<10}")

    return all_true_labels, all_predictions, accuracy, precision, recall, f1, support

def plot_confusion_matrix(y_true, y_pred, class_names, dataset_name, model_type="Predator_Aware_GOA", save_path=None):
    """
    Plot and optionally save confusion matrix
    """
    cm = confusion_matrix(y_true, y_pred)

    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=class_names, yticklabels=class_names)
    plt.title(f'Confusion Matrix - {model_type} - {dataset_name} Set', fontsize=16, fontweight='bold')
    plt.xlabel('Predicted Label', fontsize=12)
    plt.ylabel('True Label', fontsize=12)
    plt.tight_layout()

    if save_path:
        filename = f"confusion_matrix_{model_type.lower()}_{dataset_name.lower()}.png"
        plt.savefig(f"{save_path}/{filename}", dpi=300, bbox_inches='tight')
        print(f"✓ Confusion matrix saved: {filename}")

    plt.show()

    return cm

def print_classification_report(y_true, y_pred, class_names, dataset_name, model_type="Predator Aware Momentum GOA"):
    """
    Print detailed classification report using sklearn
    """
    print(f"\n{'='*50}")
    print(f"SKLEARN CLASSIFICATION REPORT - {model_type} - {dataset_name.upper()}")
    print(f"{'='*50}")
    print(classification_report(y_true, y_pred, target_names=class_names, digits=4))

# Evaluate on all datasets
results = {}
save_path = "/content/drive/MyDrive/Metal Detection Data/results/pred_aware"

print(f"\n{'='*80}")
print("PREDATOR AWARE MOMENTUM GOA MODEL EVALUATION")
print(f"{'='*80}")

# Show reported accuracies if available
if reported_accuracies:
    print("📊 EXPECTED PERFORMANCE (from model info):")
    for key, value in reported_accuracies.items():
        print(f"  {key}: {value:.4f} ({value*100:.2f}%)")
    print()

# Evaluate Training Set
print("🔍 EVALUATING TRAINING SET...")
train_true, train_pred, train_acc, train_precision, train_recall, train_f1, train_support = evaluate_model_detailed(
    model, train_loader, "Training", class_names)
results['training'] = {
    'accuracy': train_acc,
    'precision': train_precision,
    'recall': train_recall,
    'f1': train_f1,
    'support': train_support,
    'y_true': train_true,
    'y_pred': train_pred
}

# Plot confusion matrix for training set
print("\n📊 GENERATING CONFUSION MATRIX FOR TRAINING SET...")
train_cm = plot_confusion_matrix(train_true, train_pred, class_names, "Training", "Predator_Aware_GOA", save_path)

# Print sklearn classification report for training set
print_classification_report(train_true, train_pred, class_names, "Training", "Predator Aware Momentum GOA")

# Evaluate Test Set
print("🔍 EVALUATING TEST SET...")
test_true, test_pred, test_acc, test_precision, test_recall, test_f1, test_support = evaluate_model_detailed(
    model, test_loader, "Test", class_names)
results['test'] = {
    'accuracy': test_acc,
    'precision': test_precision,
    'recall': test_recall,
    'f1': test_f1,
    'support': test_support,
    'y_true': test_true,
    'y_pred': test_pred
}

# Plot confusion matrix for test set
print("\n📊 GENERATING CONFUSION MATRIX FOR TEST SET...")
test_cm = plot_confusion_matrix(test_true, test_pred, class_names, "Test", "Predator_Aware_GOA", save_path)

# Print sklearn classification report for test set
print_classification_report(test_true, test_pred, class_names, "Test", "Predator Aware Momentum GOA")

# Check model diagnostic
unique_predictions = np.unique(test_pred)
print(f"\n⚠️  MODEL DIAGNOSTIC:")
print(f"Unique predictions: {unique_predictions} (classes: {[class_names[i] for i in unique_predictions]})")
print(f"Model is predicting {len(unique_predictions)} out of {len(class_names)} classes")

# Evaluate Validation Set
print("\n🔍 EVALUATING VALIDATION SET...")
valid_true, valid_pred, valid_acc, valid_precision, valid_recall, valid_f1, valid_support = evaluate_model_detailed(
    model, valid_loader, "Validation", class_names)
results['validation'] = {
    'accuracy': valid_acc,
    'precision': valid_precision,
    'recall': valid_recall,
    'f1': valid_f1,
    'support': valid_support,
    'y_true': valid_true,
    'y_pred': valid_pred
}

# Plot confusion matrix for validation set
print("\n📊 GENERATING CONFUSION MATRIX FOR VALIDATION SET...")
valid_cm = plot_confusion_matrix(valid_true, valid_pred, class_names, "Validation", "Predator_Aware_GOA", save_path)

# Print sklearn classification report for validation set
print_classification_report(valid_true, valid_pred, class_names, "Validation", "Predator Aware Momentum GOA")

# Summary comparison
print(f"\n{'='*80}")
print("PREDATOR AWARE MOMENTUM GOA MODEL - FINAL SUMMARY COMPARISON")
print(f"{'='*80}")

# Compare with reported accuracies if available
if reported_accuracies:
    print("📋 EXPECTED vs ACTUAL PERFORMANCE:")
    print("-" * 50)
    if 'training_accuracy' in reported_accuracies:
        expected_train = reported_accuracies['training_accuracy']
        actual_train = results['training']['accuracy']
        print(f"Training:   Expected {expected_train:.4f} | Actual {actual_train:.4f} | Diff: {actual_train-expected_train:+.4f}")

    if 'validation_accuracy' in reported_accuracies:
        expected_val = reported_accuracies['validation_accuracy']
        actual_val = results['validation']['accuracy']
        print(f"Validation: Expected {expected_val:.4f} | Actual {actual_val:.4f} | Diff: {actual_val-expected_val:+.4f}")

    if 'test_accuracy' in reported_accuracies:
        expected_test = reported_accuracies['test_accuracy']
        actual_test = results['test']['accuracy']
        print(f"Test:       Expected {expected_test:.4f} | Actual {actual_test:.4f} | Diff: {actual_test-expected_test:+.4f}")

    print()

print(f"{'Dataset':<12} {'Accuracy':<10} {'Macro F1':<10} {'Weighted F1':<12}")
print("-" * 50)

# Define the order for better comparison
dataset_order = ['training', 'validation', 'test']
for dataset_name in dataset_order:
    if dataset_name in results:
        data = results[dataset_name]
        macro_f1 = np.mean(data['f1'])
        # Calculate weighted F1 properly using support information
        if 'support' in data and len(data['support']) > 0:
            weighted_f1 = np.average(data['f1'], weights=data['support'])
        else:
            weighted_f1 = macro_f1  # fallback to macro if no support info
        print(f"{dataset_name.capitalize():<12} {data['accuracy']:<10.4f} {macro_f1:<10.4f} {weighted_f1:<12.4f}")

print(f"\n{'='*80}")
print("📊 PREDATOR AWARE MOMENTUM GOA PERFORMANCE ANALYSIS:")
print(f"{'='*80}")

if 'training' in results and 'validation' in results and 'test' in results:
    train_acc = results['training']['accuracy']
    val_acc = results['validation']['accuracy']
    test_acc = results['test']['accuracy']

    print(f"Training vs Validation Gap: {train_acc - val_acc:.4f} ({((train_acc - val_acc)*100):.2f}%)")
    print(f"Training vs Test Gap:       {train_acc - test_acc:.4f} ({((train_acc - test_acc)*100):.2f}%)")
    print(f"Validation vs Test Gap:     {val_acc - test_acc:.4f} ({((val_acc - test_acc)*100):.2f}%)")

    if train_acc - val_acc > 0.05:
        print("⚠️  Large train-validation gap suggests OVERFITTING")
    if val_acc - test_acc > 0.05:
        print("⚠️  Large validation-test gap suggests possible data distribution mismatch")
    if train_acc > 0.95 and test_acc < 0.85:
        print("🚨 High training accuracy but lower test accuracy - model may be overfitting")

    print(f"\n🔍 PREDATOR AWARE MODEL QUALITY ASSESSMENT:")
    if test_acc > 0.9:
        print("✅ EXCELLENT: Test accuracy > 90%")
    elif test_acc > 0.8:
        print("✅ GOOD: Test accuracy > 80%")
    elif test_acc > 0.7:
        print("⚠️  FAIR: Test accuracy > 70%")
    else:
        print("🚨 POOR: Test accuracy < 70%")

print(f"\n🦎 PREDATOR AWARE MOMENTUM SPECIFIC INSIGHTS:")
print("Predator Aware Momentum is known for:")
print("• Enhanced avoidance of local optima")
print("• Momentum-based convergence acceleration")
print("• Adaptive exploration-exploitation balance")
print("• Better handling of complex loss landscapes")
print("• Improved robustness against getting trapped")

print(f"\n🎯 ALGORITHM CHARACTERISTICS:")
print("• Uses momentum to escape local minima")
print("• Predator awareness prevents premature convergence")
print("• Balances global search with local refinement")
print("• Often shows stable and consistent performance")

print(f"{'='*80}")
print("✅ PREDATOR AWARE MOMENTUM GOA MODEL EVALUATION COMPLETE!")
print("📁 All confusion matrices saved to:", save_path)
print(f"Loading strategy used: {loading_strategy}")
print(f"{'='*80}")
```

```python
import torch
import torch.nn as nn
from torch.utils.data import DataLoader
from torchvision import datasets, transforms
import numpy as np
from sklearn.metrics import classification_report, confusion_matrix, precision_recall_fscore_support
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
import json

# Set device
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(f"Using device: {device}")

# Define your TinyVGG model architecture
class TinyVGG_v2(nn.Module):
    def __init__(self, input_shape: int, hidden_units: int, output_shape: int) -> None:
        super().__init__()
        self.conv_block_1 = nn.Sequential(
            nn.Conv2d(in_channels=input_shape, out_channels=hidden_units, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.Conv2d(in_channels=hidden_units, out_channels=hidden_units, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2, stride=2),
        )
        self.conv_block_2 = nn.Sequential(
            nn.Conv2d(hidden_units, hidden_units, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.Conv2d(hidden_units, hidden_units, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2, stride=2),
        )
        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Linear(in_features=hidden_units * 64 * 64, out_features=output_shape),
        )

    def forward(self, x: torch.Tensor) -> torch.Tensor:
        x = self.conv_block_1(x)
        x = self.conv_block_2(x)
        x = self.classifier(x)
        return x

# Alternative TinyVGG architecture with dropout (in case it's needed)
class TinyVGG_with_Dropout(nn.Module):
    def __init__(self, input_shape: int, hidden_units: int, output_shape: int, dropout_rate: float = 0.5) -> None:
        super().__init__()
        self.conv_block_1 = nn.Sequential(
            nn.Conv2d(in_channels=input_shape, out_channels=hidden_units, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.Conv2d(in_channels=hidden_units, out_channels=hidden_units, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2, stride=2),
        )
        self.conv_block_2 = nn.Sequential(
            nn.Conv2d(hidden_units, hidden_units, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.Conv2d(hidden_units, hidden_units, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2, stride=2),
        )
        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Dropout(dropout_rate),
            nn.Linear(in_features=hidden_units * 64 * 64, out_features=output_shape),
        )

    def forward(self, x: torch.Tensor) -> torch.Tensor:
        x = self.conv_block_1(x)
        x = self.conv_block_2(x)
        x = self.classifier(x)
        return x

# Data loading setup
train_data_path = "/content/drive/MyDrive/Metal Detection Data/Processed_Data/train"
valid_data_path = "/content/drive/MyDrive/Metal Detection Data/Processed_Data/valid"
test_data_path = "/content/drive/MyDrive/Metal Detection Data/Processed_Data/test"

transform = transforms.Compose([
    transforms.Grayscale(num_output_channels=3),
    transforms.Resize((256, 256)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5])
])

train_data = datasets.ImageFolder(train_data_path, transform=transform)
valid_data = datasets.ImageFolder(valid_data_path, transform=transform)
test_data = datasets.ImageFolder(test_data_path, transform=transform)

train_loader = DataLoader(train_data, batch_size=32, shuffle=False)  # No shuffle for evaluation
valid_loader = DataLoader(valid_data, batch_size=32, shuffle=False)
test_loader = DataLoader(test_data, batch_size=32, shuffle=False)

class_names = train_data.classes
class_dict = train_data.class_to_idx
print(f"Classes: {class_names}")
print(f"Class to index mapping: {class_dict}")

# Define model file paths
base_path = "/content/drive/MyDrive/Metal Detection Data/results/random_walk"
model_path = f"{base_path}/goa_random_walk_tinyvgg_model.pth"
model_info_path = f"{base_path}/goa_model_info.json"

print(f"\n{'='*60}")
print("LOADING RANDOM WALK GOA MODEL")
print(f"{'='*60}")

# First, load the model info to understand the architecture
optimal_hidden_units = 64  # default
dropout_rate = 0.5  # default
reported_accuracies = None

try:
    with open(model_info_path, 'r') as f:
        model_info = json.load(f)
    print("✓ Model info loaded successfully")
    print(f"Model info keys: {list(model_info.keys())}")

    if 'best_parameters' in model_info:
        best_params = model_info['best_parameters']
        print(f"Best parameters: {best_params}")

        # Extract parameters
        if 'hidden_units' in best_params:
            optimal_hidden_units = best_params['hidden_units']
        if 'dropout_rate' in best_params:
            dropout_rate = best_params['dropout_rate']

    if 'accuracies' in model_info:
        reported_accuracies = model_info['accuracies']
        print(f"Reported accuracies: {reported_accuracies}")

    # Check for model type
    if 'model_type' in model_info:
        print(f"Model type: {model_info['model_type']}")

except Exception as e:
    print(f"Could not load model info: {e}")

print(f"Using hidden_units: {optimal_hidden_units}")
print(f"Using dropout_rate: {dropout_rate}")

# Load the model with multiple strategies
model_loaded = False
loading_strategy = ""

# Strategy 1: Try loading complete model with weights_only=False (for PyTorch 2.6+)
try:
    model = torch.load(model_path, map_location=device, weights_only=False)
    print("✓ Strategy 1: Loaded complete model successfully")
    loading_strategy = "complete_model"
    model_loaded = True
except Exception as e:
    print(f"Strategy 1 failed: {e}")

# Strategy 2: Try loading as state dict and reconstruct model
if not model_loaded:
    try:
        checkpoint = torch.load(model_path, map_location=device, weights_only=False)

        if isinstance(checkpoint, dict) and 'state_dict' in checkpoint:
            state_dict = checkpoint['state_dict']
        elif isinstance(checkpoint, dict):
            state_dict = checkpoint
        else:
            state_dict = checkpoint.state_dict() if hasattr(checkpoint, 'state_dict') else checkpoint

        # Auto-detect hidden_units from first conv layer
        if 'conv_block_1.0.weight' in state_dict:
            detected_hidden_units = state_dict['conv_block_1.0.weight'].shape[0]
            print(f"Detected hidden_units from weights: {detected_hidden_units}")
        else:
            detected_hidden_units = optimal_hidden_units

        # Try with dropout first (Random Walk might benefit from regularization)
        try:
            model = TinyVGG_with_Dropout(
                input_shape=3,
                hidden_units=detected_hidden_units,
                output_shape=len(class_names),
                dropout_rate=dropout_rate
            ).to(device)

            model.load_state_dict(state_dict)
            print(f"✓ Strategy 2a: Loaded model with dropout (hidden_units={detected_hidden_units})")
            loading_strategy = "weights_with_dropout"
            model_loaded = True

        except Exception:
            # Try without dropout
            model = TinyVGG_v2(
                input_shape=3,
                hidden_units=detected_hidden_units,
                output_shape=len(class_names)
            ).to(device)

            model.load_state_dict(state_dict)
            print(f"✓ Strategy 2b: Loaded model without dropout (hidden_units={detected_hidden_units})")
            loading_strategy = "weights_no_dropout"
            model_loaded = True

    except Exception as e:
        print(f"Strategy 2 failed: {e}")

# Strategy 3: Try with optimal parameters from model info
if not model_loaded:
    try:
        model = TinyVGG_v2(
            input_shape=3,
            hidden_units=optimal_hidden_units,
            output_shape=len(class_names)
        ).to(device)

        checkpoint = torch.load(model_path, map_location=device, weights_only=False)
        if isinstance(checkpoint, dict):
            model.load_state_dict(checkpoint)
        else:
            model = checkpoint

        print(f"✓ Strategy 3: Loaded model with optimal_hidden_units={optimal_hidden_units}")
        loading_strategy = "optimal_params"
        model_loaded = True

    except Exception as e:
        print(f"Strategy 3 failed: {e}")

# Strategy 4: Try multiple architectures and hidden_units combinations
if not model_loaded:
    architectures_to_try = [
        ("TinyVGG_with_Dropout", TinyVGG_with_Dropout, {"dropout_rate": dropout_rate}),
        ("TinyVGG_v2", TinyVGG_v2, {}),
    ]

    hidden_units_to_try = [optimal_hidden_units, 64, 90, 128, 32, 256]

    for arch_name, arch_class, arch_kwargs in architectures_to_try:
        for hu in hidden_units_to_try:
            try:
                model = arch_class(
                    input_shape=3,
                    hidden_units=hu,
                    output_shape=len(class_names),
                    **arch_kwargs
                ).to(device)

                checkpoint = torch.load(model_path, map_location=device, weights_only=False)
                model.load_state_dict(checkpoint)
                print(f"✓ Strategy 4: Loaded {arch_name} with hidden_units={hu}")
                loading_strategy = f"strategy4_{arch_name}_{hu}"
                model_loaded = True
                break

            except Exception:
                continue

        if model_loaded:
            break

if not model_loaded:
    print("🚨 ERROR: Could not load model with any strategy!")
    exit()

model.eval()  # Set to evaluation mode
print(f"Model loaded using: {loading_strategy}")

def evaluate_model_detailed(model, data_loader, dataset_name, class_names):
    """
    Comprehensive evaluation function that returns predictions, true labels, and metrics
    """
    model.eval()
    all_predictions = []
    all_true_labels = []
    correct = 0
    total = 0

    print(f"\n{'='*50}")
    print(f"DETAILED EVALUATION ON {dataset_name.upper()} SET")
    print(f"{'='*50}")

    with torch.no_grad():
        for batch_idx, (inputs, labels) in enumerate(tqdm(data_loader, desc=f"Evaluating {dataset_name}")):
            inputs, labels = inputs.to(device), labels.to(device)

            outputs = model(inputs)
            _, predicted = torch.max(outputs.data, 1)

            total += labels.size(0)
            correct += (predicted == labels).sum().item()

            # Store predictions and true labels
            all_predictions.extend(predicted.cpu().numpy())
            all_true_labels.extend(labels.cpu().numpy())

    # Calculate accuracy
    accuracy = correct / total

    # Calculate precision, recall, F1-score with zero_division handling
    precision, recall, f1, support = precision_recall_fscore_support(
        all_true_labels, all_predictions, average=None, labels=range(len(class_names)), zero_division=0
    )

    # Calculate macro and weighted averages
    macro_precision, macro_recall, macro_f1, _ = precision_recall_fscore_support(
        all_true_labels, all_predictions, average='macro', zero_division=0
    )

    weighted_precision, weighted_recall, weighted_f1, _ = precision_recall_fscore_support(
        all_true_labels, all_predictions, average='weighted', zero_division=0
    )

    # Print results
    print(f"\nACCURACY: {accuracy:.4f} ({accuracy*100:.2f}%)")
    print(f"\nPER-CLASS METRICS:")
    print("-" * 60)
    print(f"{'Class':<15} {'Precision':<10} {'Recall':<10} {'F1-Score':<10} {'Support':<10}")
    print("-" * 60)

    for i, class_name in enumerate(class_names):
        print(f"{class_name:<15} {precision[i]:<10.4f} {recall[i]:<10.4f} {f1[i]:<10.4f} {support[i]:<10}")

    print("-" * 60)
    print(f"{'Macro Avg':<15} {macro_precision:<10.4f} {macro_recall:<10.4f} {macro_f1:<10.4f} {sum(support):<10}")
    print(f"{'Weighted Avg':<15} {weighted_precision:<10.4f} {weighted_recall:<10.4f} {weighted_f1:<10.4f} {sum(support):<10}")

    return all_true_labels, all_predictions, accuracy, precision, recall, f1, support

def plot_confusion_matrix(y_true, y_pred, class_names, dataset_name, model_type="Random_Walk_GOA", save_path=None):
    """
    Plot and optionally save confusion matrix
    """
    cm = confusion_matrix(y_true, y_pred)

    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=class_names, yticklabels=class_names)
    plt.title(f'Confusion Matrix - {model_type} - {dataset_name} Set', fontsize=16, fontweight='bold')
    plt.xlabel('Predicted Label', fontsize=12)
    plt.ylabel('True Label', fontsize=12)
    plt.tight_layout()

    if save_path:
        filename = f"confusion_matrix_{model_type.lower()}_{dataset_name.lower()}.png"
        plt.savefig(f"{save_path}/{filename}", dpi=300, bbox_inches='tight')
        print(f"✓ Confusion matrix saved: {filename}")

    plt.show()

    return cm

def print_classification_report(y_true, y_pred, class_names, dataset_name, model_type="Random Walk GOA"):
    """
    Print detailed classification report using sklearn
    """
    print(f"\n{'='*50}")
    print(f"SKLEARN CLASSIFICATION REPORT - {model_type} - {dataset_name.upper()}")
    print(f"{'='*50}")
    print(classification_report(y_true, y_pred, target_names=class_names, digits=4))

# Evaluate on all datasets
results = {}
save_path = "/content/drive/MyDrive/Metal Detection Data/results/random_walk"

print(f"\n{'='*80}")
print("RANDOM WALK GOA MODEL EVALUATION")
print(f"{'='*80}")

# Show reported accuracies if available
if reported_accuracies:
    print("📊 EXPECTED PERFORMANCE (from model info):")
    for key, value in reported_accuracies.items():
        print(f"  {key}: {value:.4f} ({value*100:.2f}%)")
    print()

# Evaluate Training Set
print("🔍 EVALUATING TRAINING SET...")
train_true, train_pred, train_acc, train_precision, train_recall, train_f1, train_support = evaluate_model_detailed(
    model, train_loader, "Training", class_names)
results['training'] = {
    'accuracy': train_acc,
    'precision': train_precision,
    'recall': train_recall,
    'f1': train_f1,
    'support': train_support,
    'y_true': train_true,
    'y_pred': train_pred
}

# Plot confusion matrix for training set
print("\n📊 GENERATING CONFUSION MATRIX FOR TRAINING SET...")
train_cm = plot_confusion_matrix(train_true, train_pred, class_names, "Training", "Random_Walk_GOA", save_path)

# Print sklearn classification report for training set
print_classification_report(train_true, train_pred, class_names, "Training", "Random Walk GOA")

# Evaluate Test Set
print("🔍 EVALUATING TEST SET...")
test_true, test_pred, test_acc, test_precision, test_recall, test_f1, test_support = evaluate_model_detailed(
    model, test_loader, "Test", class_names)
results['test'] = {
    'accuracy': test_acc,
    'precision': test_precision,
    'recall': test_recall,
    'f1': test_f1,
    'support': test_support,
    'y_true': test_true,
    'y_pred': test_pred
}

# Plot confusion matrix for test set
print("\n📊 GENERATING CONFUSION MATRIX FOR TEST SET...")
test_cm = plot_confusion_matrix(test_true, test_pred, class_names, "Test", "Random_Walk_GOA", save_path)

# Print sklearn classification report for test set
print_classification_report(test_true, test_pred, class_names, "Test", "Random Walk GOA")

# Check model diagnostic
unique_predictions = np.unique(test_pred)
print(f"\n⚠️  MODEL DIAGNOSTIC:")
print(f"Unique predictions: {unique_predictions} (classes: {[class_names[i] for i in unique_predictions]})")
print(f"Model is predicting {len(unique_predictions)} out of {len(class_names)} classes")

# Evaluate Validation Set
print("\n🔍 EVALUATING VALIDATION SET...")
valid_true, valid_pred, valid_acc, valid_precision, valid_recall, valid_f1, valid_support = evaluate_model_detailed(
    model, valid_loader, "Validation", class_names)
results['validation'] = {
    'accuracy': valid_acc,
    'precision': valid_precision,
    'recall': valid_recall,
    'f1': valid_f1,
    'support': valid_support,
    'y_true': valid_true,
    'y_pred': valid_pred
}

# Plot confusion matrix for validation set
print("\n📊 GENERATING CONFUSION MATRIX FOR VALIDATION SET...")
valid_cm = plot_confusion_matrix(valid_true, valid_pred, class_names, "Validation", "Random_Walk_GOA", save_path)

# Print sklearn classification report for validation set
print_classification_report(valid_true, valid_pred, class_names, "Validation", "Random Walk GOA")

# Summary comparison
print(f"\n{'='*80}")
print("RANDOM WALK GOA MODEL - FINAL SUMMARY COMPARISON")
print(f"{'='*80}")

# Compare with reported accuracies if available
if reported_accuracies:
    print("📋 EXPECTED vs ACTUAL PERFORMANCE:")
    print("-" * 50)
    if 'training_accuracy' in reported_accuracies:
        expected_train = reported_accuracies['training_accuracy']
        actual_train = results['training']['accuracy']
        print(f"Training:   Expected {expected_train:.4f} | Actual {actual_train:.4f} | Diff: {actual_train-expected_train:+.4f}")

    if 'validation_accuracy' in reported_accuracies:
        expected_val = reported_accuracies['validation_accuracy']
        actual_val = results['validation']['accuracy']
        print(f"Validation: Expected {expected_val:.4f} | Actual {actual_val:.4f} | Diff: {actual_val-expected_val:+.4f}")

    if 'test_accuracy' in reported_accuracies:
        expected_test = reported_accuracies['test_accuracy']
        actual_test = results['test']['accuracy']
        print(f"Test:       Expected {expected_test:.4f} | Actual {actual_test:.4f} | Diff: {actual_test-expected_test:+.4f}")

    print()

print(f"{'Dataset':<12} {'Accuracy':<10} {'Macro F1':<10} {'Weighted F1':<12}")
print("-" * 50)

# Define the order for better comparison
dataset_order = ['training', 'validation', 'test']
for dataset_name in dataset_order:
    if dataset_name in results:
        data = results[dataset_name]
        macro_f1 = np.mean(data['f1'])
        # Calculate weighted F1 properly using support information
        if 'support' in data and len(data['support']) > 0:
            weighted_f1 = np.average(data['f1'], weights=data['support'])
        else:
            weighted_f1 = macro_f1  # fallback to macro if no support info
        print(f"{dataset_name.capitalize():<12} {data['accuracy']:<10.4f} {macro_f1:<10.4f} {weighted_f1:<12.4f}")

print(f"\n{'='*80}")
print("📊 RANDOM WALK GOA PERFORMANCE ANALYSIS:")
print(f"{'='*80}")

if 'training' in results and 'validation' in results and 'test' in results:
    train_acc = results['training']['accuracy']
    val_acc = results['validation']['accuracy']
    test_acc = results['test']['accuracy']

    print(f"Training vs Validation Gap: {train_acc - val_acc:.4f} ({((train_acc - val_acc)*100):.2f}%)")
    print(f"Training vs Test Gap:       {train_acc - test_acc:.4f} ({((train_acc - test_acc)*100):.2f}%)")
    print(f"Validation vs Test Gap:     {val_acc - test_acc:.4f} ({((val_acc - test_acc)*100):.2f}%)")

    if train_acc - val_acc > 0.05:
        print("⚠️  Large train-validation gap suggests OVERFITTING")
    if val_acc - test_acc > 0.05:
        print("⚠️  Large validation-test gap suggests possible data distribution mismatch")
    if train_acc > 0.95 and test_acc < 0.85:
        print("🚨 High training accuracy but lower test accuracy - model may be overfitting")

    print(f"\n🔍 RANDOM WALK MODEL QUALITY ASSESSMENT:")
    if test_acc > 0.9:
        print("✅ EXCELLENT: Test accuracy > 90%")
    elif test_acc > 0.8:
        print("✅ GOOD: Test accuracy > 80%")
    elif test_acc > 0.7:
        print("⚠️  FAIR: Test accuracy > 70%")
    else:
        print("🚨 POOR: Test accuracy < 70%")

print(f"\n🎲 RANDOM WALK SPECIFIC INSIGHTS:")
print("Random Walk GOA is known for:")
print("• Stochastic exploration of the search space")
print("• Good diversification capabilities")
print("• Reduced risk of premature convergence")
print("• Natural avoidance of local optima through randomness")
print("• Simple yet effective global optimization")

print(f"\n🎯 ALGORITHM CHARACTERISTICS:")
print("• Uses random steps for exploration")
print("• Balances exploration vs exploitation naturally")
print("• Often provides robust but not always optimal solutions")
print("• Good baseline performance across various problems")
print("• Variance in results due to stochastic nature")

print(f"\n📈 RANDOM WALK ADVANTAGES:")
print("• Simple implementation and understanding")
print("• No complex parameter tuning required")
print("• Naturally avoids getting stuck in local minima")
print("• Good performance on multimodal problems")
print("• Provides diverse solution exploration")

print(f"{'='*80}")
print("✅ RANDOM WALK GOA MODEL EVALUATION COMPLETE!")
print("📁 All confusion matrices saved to:", save_path)
print(f"Loading strategy used: {loading_strategy}")
print(f"{'='*80}")
```

```python
import torch
import torch.nn as nn
from torch.utils.data import DataLoader
from torchvision import datasets, transforms
import numpy as np
from sklearn.metrics import classification_report, confusion_matrix, precision_recall_fscore_support
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
import json
import pickle
import os

# Set device
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(f"Using device: {device}")

# Define your TinyVGG model architecture
class TinyVGG_v2(nn.Module):
    def __init__(self, input_shape: int, hidden_units: int, output_shape: int) -> None:
        super().__init__()
        self.conv_block_1 = nn.Sequential(
            nn.Conv2d(in_channels=input_shape, out_channels=hidden_units, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.Conv2d(in_channels=hidden_units, out_channels=hidden_units, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2, stride=2),
        )
        self.conv_block_2 = nn.Sequential(
            nn.Conv2d(hidden_units, hidden_units, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.Conv2d(hidden_units, hidden_units, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2, stride=2),
        )
        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Linear(in_features=hidden_units * 64 * 64, out_features=output_shape),
        )

    def forward(self, x: torch.Tensor) -> torch.Tensor:
        x = self.conv_block_1(x)
        x = self.conv_block_2(x)
        x = self.classifier(x)
        return x

# Alternative TinyVGG architecture with dropout (in case it's needed)
class TinyVGG_with_Dropout(nn.Module):
    def __init__(self, input_shape: int, hidden_units: int, output_shape: int, dropout_rate: float = 0.5) -> None:
        super().__init__()
        self.conv_block_1 = nn.Sequential(
            nn.Conv2d(in_channels=input_shape, out_channels=hidden_units, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.Conv2d(in_channels=hidden_units, out_channels=hidden_units, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2, stride=2),
        )
        self.conv_block_2 = nn.Sequential(
            nn.Conv2d(hidden_units, hidden_units, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.Conv2d(hidden_units, hidden_units, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2, stride=2),
        )
        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Dropout(dropout_rate),
            nn.Linear(in_features=hidden_units * 64 * 64, out_features=output_shape),
        )

    def forward(self, x: torch.Tensor) -> torch.Tensor:
        x = self.conv_block_1(x)
        x = self.conv_block_2(x)
        x = self.classifier(x)
        return x

# Data loading setup
train_data_path = "/content/drive/MyDrive/Metal Detection Data/Processed_Data/train"
valid_data_path = "/content/drive/MyDrive/Metal Detection Data/Processed_Data/valid"
test_data_path = "/content/drive/MyDrive/Metal Detection Data/Processed_Data/test"

transform = transforms.Compose([
    transforms.Grayscale(num_output_channels=3),
    transforms.Resize((256, 256)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5])
])

train_data = datasets.ImageFolder(train_data_path, transform=transform)
valid_data = datasets.ImageFolder(valid_data_path, transform=transform)
test_data = datasets.ImageFolder(test_data_path, transform=transform)

train_loader = DataLoader(train_data, batch_size=32, shuffle=False)  # No shuffle for evaluation
valid_loader = DataLoader(valid_data, batch_size=32, shuffle=False)
test_loader = DataLoader(test_data, batch_size=32, shuffle=False)

class_names = train_data.classes
class_dict = train_data.class_to_idx
print(f"Classes: {class_names}")
print(f"Class to index mapping: {class_dict}")

# Define model file paths for Roulette Wheel
base_path = "/content/drive/MyDrive/Metal Detection Data/results/roulette_wheel"
model_files = [
    f"{base_path}/my_goa_roulette_tinyvgg_complete.pth",
    f"{base_path}/goa_roulette_tinyvgg_model.pth",
    f"{base_path}/my_goa_roulette_tinyvgg_complete.pkl"
]
model_info_path = f"{base_path}/goa_model_info.json"

print(f"\n{'='*60}")
print("LOADING ROULETTE WHEEL GOA MODEL")
print(f"{'='*60}")

# First, load the model info to understand the architecture
optimal_hidden_units = 64  # default
dropout_rate = 0.5  # default
reported_accuracies = None

try:
    with open(model_info_path, 'r') as f:
        model_info = json.load(f)
    print("✓ Model info loaded successfully")
    print(f"Model info keys: {list(model_info.keys())}")

    if 'best_parameters' in model_info:
        best_params = model_info['best_parameters']
        print(f"Best parameters: {best_params}")

        # Extract parameters
        if 'hidden_units' in best_params:
            optimal_hidden_units = best_params['hidden_units']
        if 'dropout_rate' in best_params:
            dropout_rate = best_params['dropout_rate']

    if 'accuracies' in model_info:
        reported_accuracies = model_info['accuracies']
        print(f"Reported accuracies: {reported_accuracies}")

    # Check for model type
    if 'model_type' in model_info:
        print(f"Model type: {model_info['model_type']}")

except Exception as e:
    print(f"Could not load model info: {e}")

print(f"Using hidden_units: {optimal_hidden_units}")
print(f"Using dropout_rate: {dropout_rate}")

# Load the model with multiple strategies and file types
model_loaded = False
loading_strategy = ""
model_path_used = ""

# Check which model files exist
existing_files = []
for file_path in model_files:
    if os.path.exists(file_path):
        existing_files.append(file_path)
        print(f"✓ Found: {file_path}")
    else:
        print(f"✗ Missing: {file_path}")

if not existing_files:
    print("🚨 ERROR: No model files found!")
    exit()

# Try loading from each existing file
for model_path in existing_files:
    print(f"\nTrying to load from: {model_path}")

    # Strategy 1: Try loading complete model with weights_only=False
    try:
        if model_path.endswith('.pkl'):
            # Handle pickle files
            with open(model_path, 'rb') as f:
                model = pickle.load(f)
            print("✓ Strategy 1: Loaded complete model from pickle successfully")
        else:
            # Handle .pth files
            model = torch.load(model_path, map_location=device, weights_only=False)
            print("✓ Strategy 1: Loaded complete model from .pth successfully")

        loading_strategy = "complete_model"
        model_path_used = model_path
        model_loaded = True
        break
    except Exception as e:
        print(f"Strategy 1 failed: {e}")

    # Strategy 2: Try loading as state dict and reconstruct model
    if not model_loaded:
        try:
            if model_path.endswith('.pkl'):
                # Handle pickle files
                with open(model_path, 'rb') as f:
                    checkpoint = pickle.load(f)
            else:
                checkpoint = torch.load(model_path, map_location=device, weights_only=False)

            if isinstance(checkpoint, dict) and 'state_dict' in checkpoint:
                state_dict = checkpoint['state_dict']
            elif isinstance(checkpoint, dict):
                state_dict = checkpoint
            else:
                state_dict = checkpoint.state_dict() if hasattr(checkpoint, 'state_dict') else checkpoint

            # Auto-detect hidden_units from first conv layer
            if 'conv_block_1.0.weight' in state_dict:
                detected_hidden_units = state_dict['conv_block_1.0.weight'].shape[0]
                print(f"Detected hidden_units from weights: {detected_hidden_units}")
            else:
                detected_hidden_units = optimal_hidden_units

            # Try with dropout first (Roulette Wheel might benefit from regularization)
            try:
                model = TinyVGG_with_Dropout(
                    input_shape=3,
                    hidden_units=detected_hidden_units,
                    output_shape=len(class_names),
                    dropout_rate=dropout_rate
                ).to(device)

                model.load_state_dict(state_dict)
                print(f"✓ Strategy 2a: Loaded model with dropout (hidden_units={detected_hidden_units})")
                loading_strategy = "weights_with_dropout"
                model_path_used = model_path
                model_loaded = True
                break

            except Exception:
                # Try without dropout
                model = TinyVGG_v2(
                    input_shape=3,
                    hidden_units=detected_hidden_units,
                    output_shape=len(class_names)
                ).to(device)

                model.load_state_dict(state_dict)
                print(f"✓ Strategy 2b: Loaded model without dropout (hidden_units={detected_hidden_units})")
                loading_strategy = "weights_no_dropout"
                model_path_used = model_path
                model_loaded = True
                break

        except Exception as e:
            print(f"Strategy 2 failed: {e}")

    if model_loaded:
        break

# Strategy 3: Try with optimal parameters from model info
if not model_loaded:
    for model_path in existing_files:
        try:
            model = TinyVGG_v2(
                input_shape=3,
                hidden_units=optimal_hidden_units,
                output_shape=len(class_names)
            ).to(device)

            if model_path.endswith('.pkl'):
                with open(model_path, 'rb') as f:
                    checkpoint = pickle.load(f)
            else:
                checkpoint = torch.load(model_path, map_location=device, weights_only=False)

            if isinstance(checkpoint, dict):
                model.load_state_dict(checkpoint)
            else:
                model = checkpoint

            print(f"✓ Strategy 3: Loaded model with optimal_hidden_units={optimal_hidden_units}")
            loading_strategy = "optimal_params"
            model_path_used = model_path
            model_loaded = True
            break

        except Exception as e:
            print(f"Strategy 3 failed for {model_path}: {e}")

# Strategy 4: Try multiple architectures and hidden_units combinations
if not model_loaded:
    architectures_to_try = [
        ("TinyVGG_with_Dropout", TinyVGG_with_Dropout, {"dropout_rate": dropout_rate}),
        ("TinyVGG_v2", TinyVGG_v2, {}),
    ]

    hidden_units_to_try = [optimal_hidden_units, 64, 90, 128, 32, 256]

    for model_path in existing_files:
        for arch_name, arch_class, arch_kwargs in architectures_to_try:
            for hu in hidden_units_to_try:
                try:
                    model = arch_class(
                        input_shape=3,
                        hidden_units=hu,
                        output_shape=len(class_names),
                        **arch_kwargs
                    ).to(device)

                    if model_path.endswith('.pkl'):
                        with open(model_path, 'rb') as f:
                            checkpoint = pickle.load(f)
                    else:
                        checkpoint = torch.load(model_path, map_location=device, weights_only=False)

                    model.load_state_dict(checkpoint)
                    print(f"✓ Strategy 4: Loaded {arch_name} with hidden_units={hu} from {model_path}")
                    loading_strategy = f"strategy4_{arch_name}_{hu}"
                    model_path_used = model_path
                    model_loaded = True
                    break

                except Exception:
                    continue

            if model_loaded:
                break
        if model_loaded:
            break

if not model_loaded:
    print("🚨 ERROR: Could not load model with any strategy!")
    exit()

model.eval()  # Set to evaluation mode
print(f"Model loaded using: {loading_strategy}")
print(f"Model file used: {model_path_used}")

def evaluate_model_detailed(model, data_loader, dataset_name, class_names):
    """
    Comprehensive evaluation function that returns predictions, true labels, and metrics
    """
    model.eval()
    all_predictions = []
    all_true_labels = []
    correct = 0
    total = 0

    print(f"\n{'='*50}")
    print(f"DETAILED EVALUATION ON {dataset_name.upper()} SET")
    print(f"{'='*50}")

    with torch.no_grad():
        for batch_idx, (inputs, labels) in enumerate(tqdm(data_loader, desc=f"Evaluating {dataset_name}")):
            inputs, labels = inputs.to(device), labels.to(device)

            outputs = model(inputs)
            _, predicted = torch.max(outputs.data, 1)

            total += labels.size(0)
            correct += (predicted == labels).sum().item()

            # Store predictions and true labels
            all_predictions.extend(predicted.cpu().numpy())
            all_true_labels.extend(labels.cpu().numpy())

    # Calculate accuracy
    accuracy = correct / total

    # Calculate precision, recall, F1-score with zero_division handling
    precision, recall, f1, support = precision_recall_fscore_support(
        all_true_labels, all_predictions, average=None, labels=range(len(class_names)), zero_division=0
    )

    # Calculate macro and weighted averages
    macro_precision, macro_recall, macro_f1, _ = precision_recall_fscore_support(
        all_true_labels, all_predictions, average='macro', zero_division=0
    )

    weighted_precision, weighted_recall, weighted_f1, _ = precision_recall_fscore_support(
        all_true_labels, all_predictions, average='weighted', zero_division=0
    )

    # Print results
    print(f"\nACCURACY: {accuracy:.4f} ({accuracy*100:.2f}%)")
    print(f"\nPER-CLASS METRICS:")
    print("-" * 60)
    print(f"{'Class':<15} {'Precision':<10} {'Recall':<10} {'F1-Score':<10} {'Support':<10}")
    print("-" * 60)

    for i, class_name in enumerate(class_names):
        print(f"{class_name:<15} {precision[i]:<10.4f} {recall[i]:<10.4f} {f1[i]:<10.4f} {support[i]:<10}")

    print("-" * 60)
    print(f"{'Macro Avg':<15} {macro_precision:<10.4f} {macro_recall:<10.4f} {macro_f1:<10.4f} {sum(support):<10}")
    print(f"{'Weighted Avg':<15} {weighted_precision:<10.4f} {weighted_recall:<10.4f} {weighted_f1:<10.4f} {sum(support):<10}")

    return all_true_labels, all_predictions, accuracy, precision, recall, f1, support

def plot_confusion_matrix(y_true, y_pred, class_names, dataset_name, model_type="Roulette_Wheel_GOA", save_path=None):
    """
    Plot and optionally save confusion matrix
    """
    cm = confusion_matrix(y_true, y_pred)

    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=class_names, yticklabels=class_names)
    plt.title(f'Confusion Matrix - {model_type} - {dataset_name} Set', fontsize=16, fontweight='bold')
    plt.xlabel('Predicted Label', fontsize=12)
    plt.ylabel('True Label', fontsize=12)
    plt.tight_layout()

    if save_path:
        filename = f"confusion_matrix_{model_type.lower()}_{dataset_name.lower()}.png"
        plt.savefig(f"{save_path}/{filename}", dpi=300, bbox_inches='tight')
        print(f"✓ Confusion matrix saved: {filename}")

    plt.show()

    return cm

def print_classification_report(y_true, y_pred, class_names, dataset_name, model_type="Roulette Wheel GOA"):
    """
    Print detailed classification report using sklearn
    """
    print(f"\n{'='*50}")
    print(f"SKLEARN CLASSIFICATION REPORT - {model_type} - {dataset_name.upper()}")
    print(f"{'='*50}")
    print(classification_report(y_true, y_pred, target_names=class_names, digits=4))

# Evaluate on all datasets
results = {}
save_path = "/content/drive/MyDrive/Metal Detection Data/results/roulette_wheel"

print(f"\n{'='*80}")
print("ROULETTE WHEEL GOA MODEL EVALUATION")
print(f"{'='*80}")

# Show reported accuracies if available
if reported_accuracies:
    print("📊 EXPECTED PERFORMANCE (from model info):")
    for key, value in reported_accuracies.items():
        print(f"  {key}: {value:.4f} ({value*100:.2f}%)")
    print()

# Evaluate Training Set
print("🔍 EVALUATING TRAINING SET...")
train_true, train_pred, train_acc, train_precision, train_recall, train_f1, train_support = evaluate_model_detailed(
    model, train_loader, "Training", class_names)
results['training'] = {
    'accuracy': train_acc,
    'precision': train_precision,
    'recall': train_recall,
    'f1': train_f1,
    'support': train_support,
    'y_true': train_true,
    'y_pred': train_pred
}

# Plot confusion matrix for training set
print("\n📊 GENERATING CONFUSION MATRIX FOR TRAINING SET...")
train_cm = plot_confusion_matrix(train_true, train_pred, class_names, "Training", "Roulette_Wheel_GOA", save_path)

# Print sklearn classification report for training set
print_classification_report(train_true, train_pred, class_names, "Training", "Roulette Wheel GOA")

# Evaluate Test Set
print("🔍 EVALUATING TEST SET...")
test_true, test_pred, test_acc, test_precision, test_recall, test_f1, test_support = evaluate_model_detailed(
    model, test_loader, "Test", class_names)
results['test'] = {
    'accuracy': test_acc,
    'precision': test_precision,
    'recall': test_recall,
    'f1': test_f1,
    'support': test_support,
    'y_true': test_true,
    'y_pred': test_pred
}

# Plot confusion matrix for test set
print("\n📊 GENERATING CONFUSION MATRIX FOR TEST SET...")
test_cm = plot_confusion_matrix(test_true, test_pred, class_names, "Test", "Roulette_Wheel_GOA", save_path)

# Print sklearn classification report for test set
print_classification_report(test_true, test_pred, class_names, "Test", "Roulette Wheel GOA")

# Check model diagnostic
unique_predictions = np.unique(test_pred)
print(f"\n⚠️  MODEL DIAGNOSTIC:")
print(f"Unique predictions: {unique_predictions} (classes: {[class_names[i] for i in unique_predictions]})")
print(f"Model is predicting {len(unique_predictions)} out of {len(class_names)} classes")

# Evaluate Validation Set
print("\n🔍 EVALUATING VALIDATION SET...")
valid_true, valid_pred, valid_acc, valid_precision, valid_recall, valid_f1, valid_support = evaluate_model_detailed(
    model, valid_loader, "Validation", class_names)
results['validation'] = {
    'accuracy': valid_acc,
    'precision': valid_precision,
    'recall': valid_recall,
    'f1': valid_f1,
    'support': valid_support,
    'y_true': valid_true,
    'y_pred': valid_pred
}

# Plot confusion matrix for validation set
print("\n📊 GENERATING CONFUSION MATRIX FOR VALIDATION SET...")
valid_cm = plot_confusion_matrix(valid_true, valid_pred, class_names, "Validation", "Roulette_Wheel_GOA", save_path)

# Print sklearn classification report for validation set
print_classification_report(valid_true, valid_pred, class_names, "Validation", "Roulette Wheel GOA")

# Summary comparison
print(f"\n{'='*80}")
print("ROULETTE WHEEL GOA MODEL - FINAL SUMMARY COMPARISON")
print(f"{'='*80}")

# Compare with reported accuracies if available
if reported_accuracies:
    print("📋 EXPECTED vs ACTUAL PERFORMANCE:")
    print("-" * 50)
    if 'training_accuracy' in reported_accuracies:
        expected_train = reported_accuracies['training_accuracy']
        actual_train = results['training']['accuracy']
        print(f"Training:   Expected {expected_train:.4f} | Actual {actual_train:.4f} | Diff: {actual_train-expected_train:+.4f}")

    if 'validation_accuracy' in reported_accuracies:
        expected_val = reported_accuracies['validation_accuracy']
        actual_val = results['validation']['accuracy']
        print(f"Validation: Expected {expected_val:.4f} | Actual {actual_val:.4f} | Diff: {actual_val-expected_val:+.4f}")

    if 'test_accuracy' in reported_accuracies:
        expected_test = reported_accuracies['test_accuracy']
        actual_test = results['test']['accuracy']
        print(f"Test:       Expected {expected_test:.4f} | Actual {actual_test:.4f} | Diff: {actual_test-expected_test:+.4f}")

    print()

print(f"{'Dataset':<12} {'Accuracy':<10} {'Macro F1':<10} {'Weighted F1':<12}")
print("-" * 50)

# Define the order for better comparison
dataset_order = ['training', 'validation', 'test']
for dataset_name in dataset_order:
    if dataset_name in results:
        data = results[dataset_name]
        macro_f1 = np.mean(data['f1'])
        # Calculate weighted F1 properly using support information
        if 'support' in data and len(data['support']) > 0:
            weighted_f1 = np.average(data['f1'], weights=data['support'])
        else:
            weighted_f1 = macro_f1  # fallback to macro if no support info
        print(f"{dataset_name.capitalize():<12} {data['accuracy']:<10.4f} {macro_f1:<10.4f} {weighted_f1:<12.4f}")

print(f"\n{'='*80}")
print("📊 ROULETTE WHEEL GOA PERFORMANCE ANALYSIS:")
print(f"{'='*80}")

if 'training' in results and 'validation' in results and 'test' in results:
    train_acc = results['training']['accuracy']
    val_acc = results['validation']['accuracy']
    test_acc = results['test']['accuracy']

    print(f"Training vs Validation Gap: {train_acc - val_acc:.4f} ({((train_acc - val_acc)*100):.2f}%)")
    print(f"Training vs Test Gap:       {train_acc - test_acc:.4f} ({((train_acc - test_acc)*100):.2f}%)")
    print(f"Validation vs Test Gap:     {val_acc - test_acc:.4f} ({((val_acc - test_acc)*100):.2f}%)")

    if train_acc - val_acc > 0.05:
        print("⚠️  Large train-validation gap suggests OVERFITTING")
    if val_acc - test_acc > 0.05:
        print("⚠️  Large validation-test gap suggests possible data distribution mismatch")
    if train_acc > 0.95 and test_acc < 0.85:
        print("🚨 High training accuracy but lower test accuracy - model may be overfitting")

    print(f"\n🔍 ROULETTE WHEEL MODEL QUALITY ASSESSMENT:")
    if test_acc > 0.9:
        print("✅ EXCELLENT: Test accuracy > 90%")
    elif test_acc > 0.8:
        print("✅ GOOD: Test accuracy > 80%")
    elif test_acc > 0.7:
        print("⚠️  FAIR: Test accuracy > 70%")
    else:
        print("🚨 POOR: Test accuracy < 70%")

print(f"\n🎯 ROULETTE WHEEL SPECIFIC INSIGHTS:")
print("Roulette Wheel GOA is known for:")
print("• Fitness-proportionate selection mechanism")
print("• Better individuals have higher probability of selection")
print("• Maintains population diversity while favoring good solutions")
print("• Balances exploitation of good solutions with exploration")
print("• Selection pressure proportional to fitness values")

print(f"\n🎲 ALGORITHM CHARACTERISTICS:")
print("• Uses probabilistic selection based on fitness")
print("• Selection probability = fitness / sum_of_all_fitness")
print("• Higher fitness = higher selection chance")
print("• Preserves weaker solutions with low probability")
print("• Natural selection pressure toward better solutions")

print(f"\n📈 ROULETTE WHEEL ADVANTAGES:")
print("• Fitness-proportionate selection maintains diversity")
print("• Good balance between exploration and exploitation")
print("• Prevents premature convergence to local optima")
print("• Simple and intuitive selection mechanism")
print("• Adapts selection pressure based on fitness distribution")

print(f"\n🔄 SELECTION MECHANISM:")
print("• Creates cumulative fitness distribution")
print("• Random number determines selected individual")
print("• Higher fitness regions occupy more selection space")
print("• Maintains stochastic nature while biasing toward quality")
print("• Effective for multimodal optimization problems")

print(f"{'='*80}")
print("✅ ROULETTE WHEEL GOA MODEL EVALUATION COMPLETE!")
print("📁 All confusion matrices saved to:", save_path)
print(f"Loading strategy used: {loading_strategy}")
print(f"Model file used: {os.path.basename(model_path_used)}")
print(f"{'='*80}")
```

```python
import torch
import torch.nn as nn
from torch.utils.data import DataLoader
from torchvision import datasets, transforms
import numpy as np
from sklearn.metrics import classification_report, confusion_matrix, precision_recall_fscore_support
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
import json
import pickle
import os

# Set device
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(f"Using device: {device}")

# Define your TinyVGG model architecture
class TinyVGG_v2(nn.Module):
    def __init__(self, input_shape: int, hidden_units: int, output_shape: int) -> None:
        super().__init__()
        self.conv_block_1 = nn.Sequential(
            nn.Conv2d(in_channels=input_shape, out_channels=hidden_units, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.Conv2d(in_channels=hidden_units, out_channels=hidden_units, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2, stride=2),
        )
        self.conv_block_2 = nn.Sequential(
            nn.Conv2d(hidden_units, hidden_units, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.Conv2d(hidden_units, hidden_units, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2, stride=2),
        )
        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Linear(in_features=hidden_units * 64 * 64, out_features=output_shape),
        )

    def forward(self, x: torch.Tensor) -> torch.Tensor:
        x = self.conv_block_1(x)
        x = self.conv_block_2(x)
        x = self.classifier(x)
        return x

# Alternative TinyVGG architecture with dropout (in case it's needed)
class TinyVGG_with_Dropout(nn.Module):
    def __init__(self, input_shape: int, hidden_units: int, output_shape: int, dropout_rate: float = 0.5) -> None:
        super().__init__()
        self.conv_block_1 = nn.Sequential(
            nn.Conv2d(in_channels=input_shape, out_channels=hidden_units, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.Conv2d(in_channels=hidden_units, out_channels=hidden_units, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2, stride=2),
        )
        self.conv_block_2 = nn.Sequential(
            nn.Conv2d(hidden_units, hidden_units, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.Conv2d(hidden_units, hidden_units, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2, stride=2),
        )
        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Dropout(dropout_rate),
            nn.Linear(in_features=hidden_units * 64 * 64, out_features=output_shape),
        )

    def forward(self, x: torch.Tensor) -> torch.Tensor:
        x = self.conv_block_1(x)
        x = self.conv_block_2(x)
        x = self.classifier(x)
        return x

# Data loading setup
train_data_path = "/content/drive/MyDrive/Metal Detection Data/Processed_Data/train"
valid_data_path = "/content/drive/MyDrive/Metal Detection Data/Processed_Data/valid"
test_data_path = "/content/drive/MyDrive/Metal Detection Data/Processed_Data/test"

transform = transforms.Compose([
    transforms.Grayscale(num_output_channels=3),
    transforms.Resize((256, 256)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5])
])

train_data = datasets.ImageFolder(train_data_path, transform=transform)
valid_data = datasets.ImageFolder(valid_data_path, transform=transform)
test_data = datasets.ImageFolder(test_data_path, transform=transform)

train_loader = DataLoader(train_data, batch_size=32, shuffle=False)  # No shuffle for evaluation
valid_loader = DataLoader(valid_data, batch_size=32, shuffle=False)
test_loader = DataLoader(test_data, batch_size=32, shuffle=False)

class_names = train_data.classes
class_dict = train_data.class_to_idx
print(f"Classes: {class_names}")
print(f"Class to index mapping: {class_dict}")

# Define model file paths for Standard PSO
base_path = "/content/drive/MyDrive/Metal Detection Data/results/standard_pso"
model_files = [
    f"{base_path}/standard_pso_tinyvgg_model.pth"
]
model_info_path = f"{base_path}/pso_model_info.json"

print(f"\n{'='*60}")
print("LOADING STANDARD PSO GOA MODEL")
print(f"{'='*60}")

# First, load the model info to understand the architecture
optimal_hidden_units = 64  # default
dropout_rate = 0.5  # default
reported_accuracies = None
pso_params = None

try:
    with open(model_info_path, 'r') as f:
        model_info = json.load(f)
    print("✓ Model info loaded successfully")
    print(f"Model info keys: {list(model_info.keys())}")

    if 'best_parameters' in model_info:
        best_params = model_info['best_parameters']
        print(f"Best parameters: {best_params}")

        # Extract parameters
        if 'hidden_units' in best_params:
            optimal_hidden_units = best_params['hidden_units']
        if 'dropout_rate' in best_params:
            dropout_rate = best_params['dropout_rate']

    if 'accuracies' in model_info:
        reported_accuracies = model_info['accuracies']
        print(f"Reported accuracies: {reported_accuracies}")

    # Check for PSO-specific parameters
    if 'pso_parameters' in model_info:
        pso_params = model_info['pso_parameters']
        print(f"PSO parameters: {pso_params}")

    # Check for model type
    if 'model_type' in model_info:
        print(f"Model type: {model_info['model_type']}")

    # Check for optimization results
    if 'best_fitness' in model_info:
        print(f"Best fitness achieved: {model_info['best_fitness']}")

except Exception as e:
    print(f"Could not load model info: {e}")

print(f"Using hidden_units: {optimal_hidden_units}")
print(f"Using dropout_rate: {dropout_rate}")

# Load the model with multiple strategies and file types
model_loaded = False
loading_strategy = ""
model_path_used = ""

# Check which model files exist
existing_files = []
for file_path in model_files:
    if os.path.exists(file_path):
        existing_files.append(file_path)
        print(f"✓ Found: {file_path}")
    else:
        print(f"✗ Missing: {file_path}")

if not existing_files:
    print("🚨 ERROR: No model files found!")
    exit()

# Try loading from each existing file
for model_path in existing_files:
    print(f"\nTrying to load from: {model_path}")

    # Strategy 1: Try loading complete model with weights_only=False
    try:
        if model_path.endswith('.pkl'):
            # Handle pickle files
            with open(model_path, 'rb') as f:
                model = pickle.load(f)
            print("✓ Strategy 1: Loaded complete model from pickle successfully")
        else:
            # Handle .pth files
            model = torch.load(model_path, map_location=device, weights_only=False)
            print("✓ Strategy 1: Loaded complete model from .pth successfully")

        loading_strategy = "complete_model"
        model_path_used = model_path
        model_loaded = True
        break
    except Exception as e:
        print(f"Strategy 1 failed: {e}")

    # Strategy 2: Try loading as state dict and reconstruct model
    if not model_loaded:
        try:
            if model_path.endswith('.pkl'):
                # Handle pickle files
                with open(model_path, 'rb') as f:
                    checkpoint = pickle.load(f)
            else:
                checkpoint = torch.load(model_path, map_location=device, weights_only=False)

            if isinstance(checkpoint, dict) and 'state_dict' in checkpoint:
                state_dict = checkpoint['state_dict']
            elif isinstance(checkpoint, dict):
                state_dict = checkpoint
            else:
                state_dict = checkpoint.state_dict() if hasattr(checkpoint, 'state_dict') else checkpoint

            # Auto-detect hidden_units from first conv layer
            if 'conv_block_1.0.weight' in state_dict:
                detected_hidden_units = state_dict['conv_block_1.0.weight'].shape[0]
                print(f"Detected hidden_units from weights: {detected_hidden_units}")
            else:
                detected_hidden_units = optimal_hidden_units

            # Try with dropout first (PSO might benefit from regularization)
            try:
                model = TinyVGG_with_Dropout(
                    input_shape=3,
                    hidden_units=detected_hidden_units,
                    output_shape=len(class_names),
                    dropout_rate=dropout_rate
                ).to(device)

                model.load_state_dict(state_dict)
                print(f"✓ Strategy 2a: Loaded model with dropout (hidden_units={detected_hidden_units})")
                loading_strategy = "weights_with_dropout"
                model_path_used = model_path
                model_loaded = True
                break

            except Exception:
                # Try without dropout
                model = TinyVGG_v2(
                    input_shape=3,
                    hidden_units=detected_hidden_units,
                    output_shape=len(class_names)
                ).to(device)

                model.load_state_dict(state_dict)
                print(f"✓ Strategy 2b: Loaded model without dropout (hidden_units={detected_hidden_units})")
                loading_strategy = "weights_no_dropout"
                model_path_used = model_path
                model_loaded = True
                break

        except Exception as e:
            print(f"Strategy 2 failed: {e}")

    if model_loaded:
        break

# Strategy 3: Try with optimal parameters from model info
if not model_loaded:
    for model_path in existing_files:
        try:
            model = TinyVGG_v2(
                input_shape=3,
                hidden_units=optimal_hidden_units,
                output_shape=len(class_names)
            ).to(device)

            if model_path.endswith('.pkl'):
                with open(model_path, 'rb') as f:
                    checkpoint = pickle.load(f)
            else:
                checkpoint = torch.load(model_path, map_location=device, weights_only=False)

            if isinstance(checkpoint, dict):
                model.load_state_dict(checkpoint)
            else:
                model = checkpoint

            print(f"✓ Strategy 3: Loaded model with optimal_hidden_units={optimal_hidden_units}")
            loading_strategy = "optimal_params"
            model_path_used = model_path
            model_loaded = True
            break

        except Exception as e:
            print(f"Strategy 3 failed for {model_path}: {e}")

# Strategy 4: Try multiple architectures and hidden_units combinations
if not model_loaded:
    architectures_to_try = [
        ("TinyVGG_with_Dropout", TinyVGG_with_Dropout, {"dropout_rate": dropout_rate}),
        ("TinyVGG_v2", TinyVGG_v2, {}),
    ]

    hidden_units_to_try = [optimal_hidden_units, 64, 90, 128, 32, 256]

    for model_path in existing_files:
        for arch_name, arch_class, arch_kwargs in architectures_to_try:
            for hu in hidden_units_to_try:
                try:
                    model = arch_class(
                        input_shape=3,
                        hidden_units=hu,
                        output_shape=len(class_names),
                        **arch_kwargs
                    ).to(device)

                    if model_path.endswith('.pkl'):
                        with open(model_path, 'rb') as f:
                            checkpoint = pickle.load(f)
                    else:
                        checkpoint = torch.load(model_path, map_location=device, weights_only=False)

                    model.load_state_dict(checkpoint)
                    print(f"✓ Strategy 4: Loaded {arch_name} with hidden_units={hu} from {model_path}")
                    loading_strategy = f"strategy4_{arch_name}_{hu}"
                    model_path_used = model_path
                    model_loaded = True
                    break

                except Exception:
                    continue

            if model_loaded:
                break
        if model_loaded:
            break

if not model_loaded:
    print("🚨 ERROR: Could not load model with any strategy!")
    exit()

model.eval()  # Set to evaluation mode
print(f"Model loaded using: {loading_strategy}")
print(f"Model file used: {model_path_used}")

def evaluate_model_detailed(model, data_loader, dataset_name, class_names):
    """
    Comprehensive evaluation function that returns predictions, true labels, and metrics
    """
    model.eval()
    all_predictions = []
    all_true_labels = []
    correct = 0
    total = 0

    print(f"\n{'='*50}")
    print(f"DETAILED EVALUATION ON {dataset_name.upper()} SET")
    print(f"{'='*50}")

    with torch.no_grad():
        for batch_idx, (inputs, labels) in enumerate(tqdm(data_loader, desc=f"Evaluating {dataset_name}")):
            inputs, labels = inputs.to(device), labels.to(device)

            outputs = model(inputs)
            _, predicted = torch.max(outputs.data, 1)

            total += labels.size(0)
            correct += (predicted == labels).sum().item()

            # Store predictions and true labels
            all_predictions.extend(predicted.cpu().numpy())
            all_true_labels.extend(labels.cpu().numpy())

    # Calculate accuracy
    accuracy = correct / total

    # Calculate precision, recall, F1-score with zero_division handling
    precision, recall, f1, support = precision_recall_fscore_support(
        all_true_labels, all_predictions, average=None, labels=range(len(class_names)), zero_division=0
    )

    # Calculate macro and weighted averages
    macro_precision, macro_recall, macro_f1, _ = precision_recall_fscore_support(
        all_true_labels, all_predictions, average='macro', zero_division=0
    )

    weighted_precision, weighted_recall, weighted_f1, _ = precision_recall_fscore_support(
        all_true_labels, all_predictions, average='weighted', zero_division=0
    )

    # Print results
    print(f"\nACCURACY: {accuracy:.4f} ({accuracy*100:.2f}%)")
    print(f"\nPER-CLASS METRICS:")
    print("-" * 60)
    print(f"{'Class':<15} {'Precision':<10} {'Recall':<10} {'F1-Score':<10} {'Support':<10}")
    print("-" * 60)

    for i, class_name in enumerate(class_names):
        print(f"{class_name:<15} {precision[i]:<10.4f} {recall[i]:<10.4f} {f1[i]:<10.4f} {support[i]:<10}")

    print("-" * 60)
    print(f"{'Macro Avg':<15} {macro_precision:<10.4f} {macro_recall:<10.4f} {macro_f1:<10.4f} {sum(support):<10}")
    print(f"{'Weighted Avg':<15} {weighted_precision:<10.4f} {weighted_recall:<10.4f} {weighted_f1:<10.4f} {sum(support):<10}")

    return all_true_labels, all_predictions, accuracy, precision, recall, f1, support

def plot_confusion_matrix(y_true, y_pred, class_names, dataset_name, model_type="Standard_PSO_GOA", save_path=None):
    """
    Plot and optionally save confusion matrix
    """
    cm = confusion_matrix(y_true, y_pred)

    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=class_names, yticklabels=class_names)
    plt.title(f'Confusion Matrix - {model_type} - {dataset_name} Set', fontsize=16, fontweight='bold')
    plt.xlabel('Predicted Label', fontsize=12)
    plt.ylabel('True Label', fontsize=12)
    plt.tight_layout()

    if save_path:
        filename = f"confusion_matrix_{model_type.lower()}_{dataset_name.lower()}.png"
        plt.savefig(f"{save_path}/{filename}", dpi=300, bbox_inches='tight')
        print(f"✓ Confusion matrix saved: {filename}")

    plt.show()

    return cm

def print_classification_report(y_true, y_pred, class_names, dataset_name, model_type="Standard PSO GOA"):
    """
    Print detailed classification report using sklearn
    """
    print(f"\n{'='*50}")
    print(f"SKLEARN CLASSIFICATION REPORT - {model_type} - {dataset_name.upper()}")
    print(f"{'='*50}")
    print(classification_report(y_true, y_pred, target_names=class_names, digits=4))

# Evaluate on all datasets
results = {}
save_path = "/content/drive/MyDrive/Metal Detection Data/results/standard_pso"

print(f"\n{'='*80}")
print("STANDARD PSO GOA MODEL EVALUATION")
print(f"{'='*80}")

# Show reported accuracies if available
if reported_accuracies:
    print("📊 EXPECTED PERFORMANCE (from model info):")
    for key, value in reported_accuracies.items():
        print(f"  {key}: {value:.4f} ({value*100:.2f}%)")
    print()

# Show PSO parameters if available
if pso_params:
    print("⚙️  PSO OPTIMIZATION PARAMETERS:")
    for key, value in pso_params.items():
        print(f"  {key}: {value}")
    print()

# Evaluate Training Set
print("🔍 EVALUATING TRAINING SET...")
train_true, train_pred, train_acc, train_precision, train_recall, train_f1, train_support = evaluate_model_detailed(
    model, train_loader, "Training", class_names)
results['training'] = {
    'accuracy': train_acc,
    'precision': train_precision,
    'recall': train_recall,
    'f1': train_f1,
    'support': train_support,
    'y_true': train_true,
    'y_pred': train_pred
}

# Plot confusion matrix for training set
print("\n📊 GENERATING CONFUSION MATRIX FOR TRAINING SET...")
train_cm = plot_confusion_matrix(train_true, train_pred, class_names, "Training", "Standard_PSO_GOA", save_path)

# Print sklearn classification report for training set
print_classification_report(train_true, train_pred, class_names, "Training", "Standard PSO GOA")

# Evaluate Test Set
print("🔍 EVALUATING TEST SET...")
test_true, test_pred, test_acc, test_precision, test_recall, test_f1, test_support = evaluate_model_detailed(
    model, test_loader, "Test", class_names)
results['test'] = {
    'accuracy': test_acc,
    'precision': test_precision,
    'recall': test_recall,
    'f1': test_f1,
    'support': test_support,
    'y_true': test_true,
    'y_pred': test_pred
}

# Plot confusion matrix for test set
print("\n📊 GENERATING CONFUSION MATRIX FOR TEST SET...")
test_cm = plot_confusion_matrix(test_true, test_pred, class_names, "Test", "Standard_PSO_GOA", save_path)

# Print sklearn classification report for test set
print_classification_report(test_true, test_pred, class_names, "Test", "Standard PSO GOA")

# Check model diagnostic
unique_predictions = np.unique(test_pred)
print(f"\n⚠️  MODEL DIAGNOSTIC:")
print(f"Unique predictions: {unique_predictions} (classes: {[class_names[i] for i in unique_predictions]})")
print(f"Model is predicting {len(unique_predictions)} out of {len(class_names)} classes")

# Evaluate Validation Set
print("\n🔍 EVALUATING VALIDATION SET...")
valid_true, valid_pred, valid_acc, valid_precision, valid_recall, valid_f1, valid_support = evaluate_model_detailed(
    model, valid_loader, "Validation", class_names)
results['validation'] = {
    'accuracy': valid_acc,
    'precision': valid_precision,
    'recall': valid_recall,
    'f1': valid_f1,
    'support': valid_support,
    'y_true': valid_true,
    'y_pred': valid_pred
}

# Plot confusion matrix for validation set
print("\n📊 GENERATING CONFUSION MATRIX FOR VALIDATION SET...")
valid_cm = plot_confusion_matrix(valid_true, valid_pred, class_names, "Validation", "Standard_PSO_GOA", save_path)

# Print sklearn classification report for validation set
print_classification_report(valid_true, valid_pred, class_names, "Validation", "Standard PSO GOA")

# Summary comparison
print(f"\n{'='*80}")
print("STANDARD PSO GOA MODEL - FINAL SUMMARY COMPARISON")
print(f"{'='*80}")

# Compare with reported accuracies if available
if reported_accuracies:
    print("📋 EXPECTED vs ACTUAL PERFORMANCE:")
    print("-" * 50)
    if 'training_accuracy' in reported_accuracies:
        expected_train = reported_accuracies['training_accuracy']
        actual_train = results['training']['accuracy']
        print(f"Training:   Expected {expected_train:.4f} | Actual {actual_train:.4f} | Diff: {actual_train-expected_train:+.4f}")

    if 'validation_accuracy' in reported_accuracies:
        expected_val = reported_accuracies['validation_accuracy']
        actual_val = results['validation']['accuracy']
        print(f"Validation: Expected {expected_val:.4f} | Actual {actual_val:.4f} | Diff: {actual_val-expected_val:+.4f}")

    if 'test_accuracy' in reported_accuracies:
        expected_test = reported_accuracies['test_accuracy']
        actual_test = results['test']['accuracy']
        print(f"Test:       Expected {expected_test:.4f} | Actual {actual_test:.4f} | Diff: {actual_test-expected_test:+.4f}")

    print()

print(f"{'Dataset':<12} {'Accuracy':<10} {'Macro F1':<10} {'Weighted F1':<12}")
print("-" * 50)

# Define the order for better comparison
dataset_order = ['training', 'validation', 'test']
for dataset_name in dataset_order:
    if dataset_name in results:
        data = results[dataset_name]
        macro_f1 = np.mean(data['f1'])
        # Calculate weighted F1 properly using support information
        if 'support' in data and len(data['support']) > 0:
            weighted_f1 = np.average(data['f1'], weights=data['support'])
        else:
            weighted_f1 = macro_f1  # fallback to macro if no support info
        print(f"{dataset_name.capitalize():<12} {data['accuracy']:<10.4f} {macro_f1:<10.4f} {weighted_f1:<12.4f}")

print(f"\n{'='*80}")
print("📊 STANDARD PSO GOA PERFORMANCE ANALYSIS:")
print(f"{'='*80}")

if 'training' in results and 'validation' in results and 'test' in results:
    train_acc = results['training']['accuracy']
    val_acc = results['validation']['accuracy']
    test_acc = results['test']['accuracy']

    print(f"Training vs Validation Gap: {train_acc - val_acc:.4f} ({((train_acc - val_acc)*100):.2f}%)")
    print(f"Training vs Test Gap:       {train_acc - test_acc:.4f} ({((train_acc - test_acc)*100):.2f}%)")
    print(f"Validation vs Test Gap:     {val_acc - test_acc:.4f} ({((val_acc - test_acc)*100):.2f}%)")

    if train_acc - val_acc > 0.05:
        print("⚠️  Large train-validation gap suggests OVERFITTING")
    if val_acc - test_acc > 0.05:
        print("⚠️  Large validation-test gap suggests possible data distribution mismatch")
    if train_acc > 0.95 and test_acc < 0.85:
        print("🚨 High training accuracy but lower test accuracy - model may be overfitting")

    print(f"\n🔍 STANDARD PSO MODEL QUALITY ASSESSMENT:")
    if test_acc > 0.9:
        print("✅ EXCELLENT: Test accuracy > 90%")
    elif test_acc > 0.8:
        print("✅ GOOD: Test accuracy > 80%")
    elif test_acc > 0.7:
        print("⚠️  FAIR: Test accuracy > 70%")
    else:
        print("🚨 POOR: Test accuracy < 70%")

print(f"\n🌟 STANDARD PSO SPECIFIC INSIGHTS:")
print("Particle Swarm Optimization (PSO) is known for:")
print("• Swarm intelligence inspired by bird flocking behavior")
print("• Each particle maintains position and velocity vectors")
print("• Social learning through personal and global best experiences")
print("• Fast convergence toward promising regions of search space")
print("• Balance between exploration (diversity) and exploitation (convergence)")

print(f"\n⚙️  PSO ALGORITHM CHARACTERISTICS:")
print("• Particles update velocity based on:")
print("  - Inertia: Current velocity * inertia weight")
print("  - Personal best: Cognitive component * random * (pbest - position)")
print("  - Global best: Social component * random * (gbest - position)")
print("• Position updated as: position = position + velocity")
print("• Typically converges faster than genetic algorithms")
print("• Good at fine-tuning solutions near optima")

print(f"\n📈 STANDARD PSO ADVANTAGES:")
print("• Simple implementation with few parameters to tune")
print("• Fast convergence due to information sharing")
print("• Good exploitation of promising search regions")
print("• Memory of personal and global best solutions")
print("• Effective for continuous optimization problems")
print("• Natural balance between individual and social learning")

print(f"\n🔄 PSO OPTIMIZATION PROCESS:")
print("• Initialize random particle positions and velocities")
print("• Evaluate fitness of each particle")
print("• Update personal best (pbest) for each particle")
print("• Update global best (gbest) for the swarm")
print("• Update velocities using PSO velocity equation")
print("• Update positions and repeat until convergence")

print(f"\n🎯 PSO PARAMETER INFLUENCE:")
if pso_params:
    print("• Parameters used in this optimization:")
    for param, value in pso_params.items():
        if param == 'c1':
            print(f"  - c1 (cognitive): {value} - personal best influence")
        elif param == 'c2':
            print(f"  - c2 (social): {value} - global best influence")
        elif param == 'w':
            print(f"  - w (inertia): {value} - momentum preservation")
        elif param == 'particles':
            print(f"  - population size: {value} particles")
        elif param == 'iterations':
            print(f"  - max iterations: {value}")
        else:
            print(f"  - {param}: {value}")
else:
    print("• Typical PSO parameters:")
    print("  - c1 (cognitive): ~2.0 - personal exploration")
    print("  - c2 (social): ~2.0 - swarm collaboration")
    print("  - w (inertia): 0.4-0.9 - velocity damping")
    print("  - Population: 20-50 particles typically")

print(f"{'='*80}")
print("✅ STANDARD PSO GOA MODEL EVALUATION COMPLETE!")
print("📁 All confusion matrices saved to:", save_path)
print(f"Loading strategy used: {loading_strategy}")
print(f"Model file used: {os.path.basename(model_path_used)}")
print(f"{'='*80}")
```

```python
import torch
import torch.nn as nn
from torch.utils.data import DataLoader
from torchvision import datasets, transforms
import numpy as np
from sklearn.metrics import classification_report, confusion_matrix, precision_recall_fscore_support
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
import json
import pickle
import os

# Set device
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(f"Using device: {device}")

# CORRECTED TinyVGG architecture with BatchNorm (matching the saved model)
class TinyVGG_with_BatchNorm(nn.Module):
    def __init__(self, input_shape: int, hidden_units: int, output_shape: int) -> None:
        super().__init__()
        self.conv_block_1 = nn.Sequential(
            nn.Conv2d(in_channels=input_shape, out_channels=hidden_units, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(hidden_units),  # Index 1
            nn.ReLU(),                     # Index 2
            nn.Conv2d(in_channels=hidden_units, out_channels=hidden_units, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(hidden_units),  # Index 4
            nn.ReLU(),                     # Index 5
            nn.MaxPool2d(kernel_size=2, stride=2),  # Index 6
        )
        self.conv_block_2 = nn.Sequential(
            nn.Conv2d(hidden_units, hidden_units, kernel_size=3, padding=1),
            nn.BatchNorm2d(hidden_units),  # Index 1
            nn.ReLU(),                     # Index 2
            nn.Conv2d(hidden_units, hidden_units, kernel_size=3, padding=1),
            nn.BatchNorm2d(hidden_units),  # Index 4
            nn.ReLU(),                     # Index 5
            nn.MaxPool2d(kernel_size=2, stride=2),  # Index 6
        )
        self.classifier = nn.Sequential(
            nn.Flatten(),                  # Index 0
            nn.Dropout(0.5),              # Index 1
            nn.Linear(in_features=hidden_units * 64 * 64, out_features=output_shape),  # Index 2
        )

    def forward(self, x: torch.Tensor) -> torch.Tensor:
        x = self.conv_block_1(x)
        x = self.conv_block_2(x)
        x = self.classifier(x)
        return x

# Alternative without dropout in classifier
class TinyVGG_BatchNorm_NoDropout(nn.Module):
    def __init__(self, input_shape: int, hidden_units: int, output_shape: int) -> None:
        super().__init__()
        self.conv_block_1 = nn.Sequential(
            nn.Conv2d(in_channels=input_shape, out_channels=hidden_units, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(hidden_units),
            nn.ReLU(),
            nn.Conv2d(in_channels=hidden_units, out_channels=hidden_units, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(hidden_units),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2, stride=2),
        )
        self.conv_block_2 = nn.Sequential(
            nn.Conv2d(hidden_units, hidden_units, kernel_size=3, padding=1),
            nn.BatchNorm2d(hidden_units),
            nn.ReLU(),
            nn.Conv2d(hidden_units, hidden_units, kernel_size=3, padding=1),
            nn.BatchNorm2d(hidden_units),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2, stride=2),
        )
        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Linear(in_features=hidden_units * 64 * 64, out_features=output_shape),
        )

    def forward(self, x: torch.Tensor) -> torch.Tensor:
        x = self.conv_block_1(x)
        x = self.conv_block_2(x)
        x = self.classifier(x)
        return x

# Data loading setup
train_data_path = "/content/drive/MyDrive/Metal Detection Data/Processed_Data/train"
valid_data_path = "/content/drive/MyDrive/Metal Detection Data/Processed_Data/valid"
test_data_path = "/content/drive/MyDrive/Metal Detection Data/Processed_Data/test"

transform = transforms.Compose([
    transforms.Grayscale(num_output_channels=3),
    transforms.Resize((256, 256)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5])
])

train_data = datasets.ImageFolder(train_data_path, transform=transform)
valid_data = datasets.ImageFolder(valid_data_path, transform=transform)
test_data = datasets.ImageFolder(test_data_path, transform=transform)

train_loader = DataLoader(train_data, batch_size=32, shuffle=False)
valid_loader = DataLoader(valid_data, batch_size=32, shuffle=False)
test_loader = DataLoader(test_data, batch_size=32, shuffle=False)

class_names = train_data.classes
class_dict = train_data.class_to_idx
print(f"Classes: {class_names}")
print(f"Class to index mapping: {class_dict}")

# Fix the JSON file and load model info
base_path = "/content/drive/MyDrive/Metal Detection Data/results/genetic_algorithm"
model_info_path = f"{base_path}/standard_ga_tinyvgg_info.json"

print(f"\n{'='*60}")
print("LOADING GENETIC ALGORITHM MODEL (CORRECTED)")
print(f"{'='*60}")

# Create corrected JSON content based on diagnostic
model_info = {
    "model_name": "standard_ga_tinyvgg",
    "algorithm_used": "Genetic Algorithm",
    "model_architecture": "TinyVGG_with_BatchNorm",
    "input_channels": 3,
    "hidden_units": 109,
    "output_classes": 6,
    "use_batch_norm": True,
    "dropout_rate": 0.5
}

print("✓ Model info reconstructed from diagnostic")
print(f"Architecture: {model_info['model_architecture']}")
print(f"Hidden units: {model_info['hidden_units']}")
print(f"BatchNorm: {model_info['use_batch_norm']}")

# Load the model with correct architecture
weights_path = f"{base_path}/standard_ga_tinyvgg_weights.pth"

print(f"\nLoading model from: {weights_path}")

try:
    # Load the state dict
    state_dict = torch.load(weights_path, map_location=device, weights_only=False)
    print(f"✓ State dict loaded successfully")
    print(f"Parameters in state dict: {len(state_dict)}")

    # Verify the architecture matches
    hidden_units = state_dict['conv_block_1.0.weight'].shape[0]
    output_classes = state_dict['classifier.2.weight'].shape[0]  # Note: index 2 due to dropout

    print(f"Detected hidden_units: {hidden_units}")
    print(f"Detected output_classes: {output_classes}")

    # Try different architectures to match the state dict
    architectures_to_try = [
        ("TinyVGG_with_BatchNorm", TinyVGG_with_BatchNorm),
        ("TinyVGG_BatchNorm_NoDropout", TinyVGG_BatchNorm_NoDropout),
    ]

    model_loaded = False

    for arch_name, arch_class in architectures_to_try:
        try:
            print(f"\nTrying {arch_name}...")
            model = arch_class(
                input_shape=3,
                hidden_units=hidden_units,
                output_shape=output_classes
            ).to(device)

            # Load state dict
            missing_keys, unexpected_keys = model.load_state_dict(state_dict, strict=False)

            if not missing_keys:  # No missing keys means perfect match
                print(f"✅ {arch_name} loaded successfully!")
                loading_strategy = arch_name
                model_loaded = True
                break
            else:
                print(f"Missing keys in {arch_name}: {len(missing_keys)}")

        except Exception as e:
            print(f"{arch_name} failed: {e}")

    if not model_loaded:
        print("🚨 Could not find matching architecture, using best attempt...")
        model = TinyVGG_with_BatchNorm(
            input_shape=3,
            hidden_units=hidden_units,
            output_shape=output_classes
        ).to(device)
        model.load_state_dict(state_dict, strict=False)
        loading_strategy = "TinyVGG_with_BatchNorm_forced"
        model_loaded = True

except Exception as e:
    print(f"🚨 ERROR: Could not load model: {e}")
    exit()

model.eval()
print(f"\n✅ Model loaded successfully!")
print(f"Architecture used: {loading_strategy}")
print(f"Hidden units: {hidden_units}")
print(f"Output classes: {output_classes}")

def evaluate_model_detailed(model, data_loader, dataset_name, class_names):
    """Comprehensive evaluation function"""
    model.eval()
    all_predictions = []
    all_true_labels = []
    correct = 0
    total = 0

    print(f"\n{'='*50}")
    print(f"DETAILED EVALUATION ON {dataset_name.upper()} SET")
    print(f"{'='*50}")

    with torch.no_grad():
        for batch_idx, (inputs, labels) in enumerate(tqdm(data_loader, desc=f"Evaluating {dataset_name}")):
            inputs, labels = inputs.to(device), labels.to(device)

            outputs = model(inputs)
            _, predicted = torch.max(outputs.data, 1)

            total += labels.size(0)
            correct += (predicted == labels).sum().item()

            all_predictions.extend(predicted.cpu().numpy())
            all_true_labels.extend(labels.cpu().numpy())

    accuracy = correct / total
    precision, recall, f1, support = precision_recall_fscore_support(
        all_true_labels, all_predictions, average=None, labels=range(len(class_names)), zero_division=0
    )

    macro_precision, macro_recall, macro_f1, _ = precision_recall_fscore_support(
        all_true_labels, all_predictions, average='macro', zero_division=0
    )
    weighted_precision, weighted_recall, weighted_f1, _ = precision_recall_fscore_support(
        all_true_labels, all_predictions, average='weighted', zero_division=0
    )

    print(f"\nACCURACY: {accuracy:.4f} ({accuracy*100:.2f}%)")
    print(f"\nPER-CLASS METRICS:")
    print("-" * 60)
    print(f"{'Class':<15} {'Precision':<10} {'Recall':<10} {'F1-Score':<10} {'Support':<10}")
    print("-" * 60)

    for i, class_name in enumerate(class_names):
        print(f"{class_name:<15} {precision[i]:<10.4f} {recall[i]:<10.4f} {f1[i]:<10.4f} {support[i]:<10}")

    print("-" * 60)
    print(f"{'Macro Avg':<15} {macro_precision:<10.4f} {macro_recall:<10.4f} {macro_f1:<10.4f} {sum(support):<10}")
    print(f"{'Weighted Avg':<15} {weighted_precision:<10.4f} {weighted_recall:<10.4f} {weighted_f1:<10.4f} {sum(support):<10}")

    return all_true_labels, all_predictions, accuracy, precision, recall, f1, support

def plot_confusion_matrix(y_true, y_pred, class_names, dataset_name, model_type="Genetic_Algorithm", save_path=None):
    """Plot confusion matrix"""
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=class_names, yticklabels=class_names)
    plt.title(f'Confusion Matrix - {model_type} - {dataset_name} Set', fontsize=16, fontweight='bold')
    plt.xlabel('Predicted Label', fontsize=12)
    plt.ylabel('True Label', fontsize=12)
    plt.tight_layout()

    if save_path:
        filename = f"confusion_matrix_{model_type.lower()}_{dataset_name.lower()}.png"
        plt.savefig(f"{save_path}/{filename}", dpi=300, bbox_inches='tight')
        print(f"✓ Confusion matrix saved: {filename}")
    plt.show()
    return cm

def print_classification_report(y_true, y_pred, class_names, dataset_name, model_type="Genetic Algorithm"):
    """Print detailed classification report"""
    print(f"\n{'='*50}")
    print(f"SKLEARN CLASSIFICATION REPORT - {model_type} - {dataset_name.upper()}")
    print(f"{'='*50}")
    print(classification_report(y_true, y_pred, target_names=class_names, digits=4))

# Run comprehensive evaluation
results = {}
save_path = base_path

print(f"\n{'='*80}")
print("CORRECTED GENETIC ALGORITHM MODEL EVALUATION")
print(f"{'='*80}")

# Evaluate Training Set
print("🔍 EVALUATING TRAINING SET...")
train_true, train_pred, train_acc, train_precision, train_recall, train_f1, train_support = evaluate_model_detailed(
    model, train_loader, "Training", class_names)
results['training'] = {
    'accuracy': train_acc, 'precision': train_precision, 'recall': train_recall,
    'f1': train_f1, 'support': train_support, 'y_true': train_true, 'y_pred': train_pred
}

print("\n📊 GENERATING CONFUSION MATRIX FOR TRAINING SET...")
train_cm = plot_confusion_matrix(train_true, train_pred, class_names, "Training", "Genetic_Algorithm", save_path)
print_classification_report(train_true, train_pred, class_names, "Training", "Genetic Algorithm")

# Evaluate Test Set
print("🔍 EVALUATING TEST SET...")
test_true, test_pred, test_acc, test_precision, test_recall, test_f1, test_support = evaluate_model_detailed(
    model, test_loader, "Test", class_names)
results['test'] = {
    'accuracy': test_acc, 'precision': test_precision, 'recall': test_recall,
    'f1': test_f1, 'support': test_support, 'y_true': test_true, 'y_pred': test_pred
}

print("\n📊 GENERATING CONFUSION MATRIX FOR TEST SET...")
test_cm = plot_confusion_matrix(test_true, test_pred, class_names, "Test", "Genetic_Algorithm", save_path)
print_classification_report(test_true, test_pred, class_names, "Test", "Genetic Algorithm")

# Evaluate Validation Set
print("\n🔍 EVALUATING VALIDATION SET...")
valid_true, valid_pred, valid_acc, valid_precision, valid_recall, valid_f1, valid_support = evaluate_model_detailed(
    model, valid_loader, "Validation", class_names)
results['validation'] = {
    'accuracy': valid_acc, 'precision': valid_precision, 'recall': valid_recall,
    'f1': valid_f1, 'support': valid_support, 'y_true': valid_true, 'y_pred': valid_pred
}

print("\n📊 GENERATING CONFUSION MATRIX FOR VALIDATION SET...")
valid_cm = plot_confusion_matrix(valid_true, valid_pred, class_names, "Validation", "Genetic_Algorithm", save_path)
print_classification_report(valid_true, valid_pred, class_names, "Validation", "Genetic Algorithm")

# Model diagnostics
unique_predictions = np.unique(test_pred)
print(f"\n⚠️  MODEL DIAGNOSTIC:")
print(f"Unique predictions: {unique_predictions} (classes: {[class_names[i] for i in unique_predictions]})")
print(f"Model is predicting {len(unique_predictions)} out of {len(class_names)} classes")

# Final Summary
print(f"\n{'='*80}")
print("GENETIC ALGORITHM MODEL - CORRECTED FINAL SUMMARY")
print(f"{'='*80}")

print(f"{'Dataset':<12} {'Accuracy':<10} {'Macro F1':<10} {'Weighted F1':<12}")
print("-" * 50)

for dataset_name in ['training', 'validation', 'test']:
    if dataset_name in results:
        data = results[dataset_name]
        macro_f1 = np.mean(data['f1'])
        weighted_f1 = np.average(data['f1'], weights=data['support']) if len(data['support']) > 0 else macro_f1
        print(f"{dataset_name.capitalize():<12} {data['accuracy']:<10.4f} {macro_f1:<10.4f} {weighted_f1:<12.4f}")

if 'training' in results and 'validation' in results and 'test' in results:
    train_acc = results['training']['accuracy']
    val_acc = results['validation']['accuracy']
    test_acc = results['test']['accuracy']

    print(f"\n📊 GENETIC ALGORITHM PERFORMANCE ANALYSIS:")
    print(f"Training vs Validation Gap: {train_acc - val_acc:.4f} ({((train_acc - val_acc)*100):.2f}%)")
    print(f"Training vs Test Gap:       {train_acc - test_acc:.4f} ({((train_acc - test_acc)*100):.2f}%)")
    print(f"Validation vs Test Gap:     {val_acc - test_acc:.4f} ({((val_acc - test_acc)*100):.2f}%)")

    if train_acc - val_acc > 0.05:
        print("⚠️  Large train-validation gap suggests OVERFITTING")
    if val_acc - test_acc > 0.05:
        print("⚠️  Large validation-test gap suggests possible data distribution mismatch")

    print(f"\n🔍 GENETIC ALGORITHM MODEL QUALITY:")
    if test_acc > 0.9:
        print("✅ EXCELLENT: Test accuracy > 90%")
    elif test_acc > 0.8:
        print("✅ GOOD: Test accuracy > 80%")
    elif test_acc > 0.7:
        print("⚠️  FAIR: Test accuracy > 70%")
    else:
        print("🚨 POOR: Test accuracy < 70%")

print(f"\n🧬 GENETIC ALGORITHM INSIGHTS:")
print("This GA model uses BatchNorm layers for better training stability:")
print("• BatchNorm normalizes layer inputs for faster convergence")
print("• Reduces internal covariate shift during training")
print("• Acts as implicit regularization, reducing overfitting")
print("• 109 hidden units: Non-standard size discovered through evolution")
print("• Demonstrates GA's ability to explore unconventional architectures")

print(f"\n{'='*80}")
print("✅ CORRECTED GENETIC ALGORITHM MODEL EVALUATION COMPLETE!")
print(f"Architecture: {loading_strategy}")
print(f"Hidden Units: {hidden_units} (GA-optimized)")
print(f"BatchNorm: Enabled for better training dynamics")
print(f"{'='*80}")
```
